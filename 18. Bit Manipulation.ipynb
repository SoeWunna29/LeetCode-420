###***1. Single Number (Easy)***
https://leetcode.com/problems/single-number/

```python
from typing import List

class Solution:
  def singleNumber(self, nums: List[int]) -> int:
    res = 0
    for n in nums:
      res = n ^ res # XOR (return true if both inputs are different)
    return res
```

###***2. Number of 1 Bits (Easy)***
https://leetcode.com/problems/number-of-1-bits/

```python
class Solution:
  def hammingWeight(self, n: int) -> int:
    res = 0

    while n:
      res += n % 2
      n = n >> 1
    return res
```

```python
class Solution:
  def hammingWeight(self, n: int) -> int:
    res = 0

    while n:
      n &= (n - 1)
      res += 1
    return res
```

###***3. Counting Bits (Easy)***
https://leetcode.com/problems/counting-bits/

```python
class Solution:
  def countBits(self, n: int) -> int:
    dp = [0] * (n + 1)
    offset = 1

    for i in range(1, n + 1):
      if offset * 2 == i:
        offset = i
      dp[i] = 1 + dp[i - offset]

    return dp
```

###***4. Reverse Bits (Easy)***
https://leetcode.com/problems/reverse-bits/

```python
class Solution:
  def reverseBits(self, n: int) -> int:
    res = 0

    for i in range(32):
      bit = (n >> i) & 1
      res = res | (bit << (31 - i))
    return res
```

###***5. Missing Number (Easy)***
https://leetcode.com/problems/missing-number/

```python
from typing import List

class Solution:
  def missingNumber(self, nums: List[int]) -> int:
    res = len(nums)

    for i in range(len(nums)):
      res += (i - nums[i])

    return res
```

###***6. Shuffle the Array (Easy)***
https://leetcode.com/problems/shuffle-the-array/

```python
from typing import List

class Solution:
  def shuffle(self, nums: List[int], n: int) -> List[int]:
    res = []
    for i in range(n):
      res.append(nums[i])
      res.append(nums[i + n])
    return res
```

```python
from typing import List

class Solution:
  def shuffle(self, nums: List[int], n: int) -> List[int]:
    for i in range(n):
      nums[i] = nums[i] << 10
      nums[i] = nums[i] | nums[i + n]

    j = 2 * n - 1
    for i in range(n - 1, -1, -1):
      y = nums[i] & (2 ** 10 - 1)
      x = nums[i] >> 10
      nums[j] = y
      nums[j - 1] = x
      j -= 2
    return nums
```

###***7. Add to Array-Form of Integer (Easy)***
https://leetcode.com/problems/add-to-array-form-of-integer/

```python
from typing import List

class Solution:
  def addToArrayForm(self, num: List[int], k: int) -> List[int]:
    num.reverse()
    i = 0
    while k:
      digit = k % 10 # 43 % 10 = 3, we get last digit
      if i < len(num):
        num[i] += digit
      else: # i is out of bound
        num.append(digit)

      carry = num[i] // 10 # Getting carry if it exists
      num[i] = num[i] % 10

      k = k // 10 # Chop off the one space
      k += carry
      i += 1

    num.reverse()
    return num
```

###***8. Sum of Two Integers (Medium)***
https://leetcode.com/problems/add-to-array-form-of-integer/

```python
def get_sum(a, b):
  # Iterate until there is no carry
  while b != 0:
    # Carry now contains common set bits of a and b
    carry = a & b

    # Sum of bits of a and b where at least one of the bits is not set
    a = a ^ b

    # Carry is shifted by one so that adding it to a gives the required sum
    b = carry << 1

  return a

# Example usage:
print(get_sum(1, 2))  # Output: 3
print(get_sum(2, 3))  # Output: 5
```

###***10. Add Binary (Easy)***
https://leetcode.com/problems/add-binary/

```python
class Solution:
  def addBinary(self, a: str, b: str) -> str:
    res = ""
    carry = 0

    a, b = a[::-1], b[::-1]

    for i in range(max(len(a), len(b))):
      digitA = ord(a[i]) - ord("0") if i < len(a) else 0
      digitB = ord(b[i]) - ord("0") if i < len(b) else 0

      total = digitA + digitB + carry
      char = str(total % 2) # Binary mod
      res = char + res
      carry = total // 2

    if carry:
      res = "1" + res

    return res
```
