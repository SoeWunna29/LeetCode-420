###***1. Unique Paths (Medium)***
https://leetcode.com/problems/unique-paths/

```python
class Solution:
  def uniquePaths(self, m: int, n: int) -> int:
    row = [1] * n

    for i in range(m - 1):
      newRow = [1] * n
      for j in range(n - 2, -1, -1):
        newRow[j] = newRow[j + 1] + row[j] # right cell + bottom cell
      row = newRow
    return row[0]
```

###***2. Unique Paths II (Medium)***
https://leetcode.com/problems/unique-paths-ii/

```python
from typing import List

class Solution:
  def uniquePathsWithObstacles(self, grid: List[List[int]]) -> int:
    M, N = len(grid), len(grid[0])
    dp = [0] * N # Only using 1-D instead of 2-D to optimize space
    dp[N - 1] = 1 # Setting last val in array to 1

    # Time: O(N*M), Space: O(N)
    for r in reversed(range(M)):
      for c in reversed(range(N)):
        if grid[r][c]: # Obstacle
          dp[c] = 0
        elif c + 1 < N: # Checking if c + 1 is in bound
          dp[c] = dp[c] + dp[c + 1] # bot val (cur val in this case) + right val
    return dp[0]
```

###***3. Longest Common Subsequence (Medium)***
https://leetcode.com/problems/longest-common-subsequence/

```python
class Solution:
  def longestCommonSubsequence(self, text1: str, text2: str) -> int:
    # Building 2d 2DP matrix, one extra col and row
    dp = [[0 for j in range(len(text2) + 1)] for i in range(len(text1) + 1)]

    for i in range(len(text1) -1, -1, -1): # Bottom up
      for j in range(len(text2) -1, -1, -1):
        if text1[i] == text2[j]: # if these values are equal
          dp[i][j] = 1 + dp[i + 1][j + 1] # [i + 1][j + 1] is diagonal value
        else:
          dp[i][j] = max(dp[i][j + 1], dp[i + 1][j]) # take max of right and below value

    return dp[0][0] # top left contains result
```

###***5. Last Stone Weight II (Medium)***
https://leetcode.com/problems/last-stone-weight-ii/

```python
from typing import List

class Solution:
  def lastStoneWeightII(self, stones: List[int]) -> int:
    # Memoization
    stoneSum = sum(stones)
    target = ceil(stoneSum / 2) # Trick, split stones into 2 piles

    def dfs(i, total):
      if total >= target or i == len(stones):
        return abs(total - (stoneSum - total)) # remaining stone = total - difference between two piles
      if (i, total) in dp:
        return dp[(i, total)]

      dp[(i, total)] = min(dfs(i + 1, total),
                           dfs(i + 1, total + stones[i])) # Not include, include
      return dp[(i, total)]

    dp = {}
    return dfs(0, 0)

# Bounded Knapsack
```

###***6. Best Time to Buy And Sell Stock With Cooldown (Medium)***
https://leetcode.com/problems/best-time-to-buy-and-sell-stock-with-cooldown/

```python
from typing import List

class Solution:
  def maxProfit(self, prices: List[int]) -> int:
    # State: Buying or Selling?
    # If Buy -> i + 1
    # If Sell -> i + 2 , need to skip one cooldown day

    dp = {}  # key=(i, buying) val=max_profit

    def dfs(i, buying):
      if i >= len(prices):
        return 0
      if (i, buying) in dp:
        return dp[(i, buying)]

      cooldown = dfs(i + 1, buying)
      if buying:
        buy = dfs(i + 1, not buying) - prices[i] # need to subtract purchase price
        dp[(i, buying)] = max(buy, cooldown)
      else:
        sell = dfs(i + 2, not buying) + prices[i] # add sold price
        dp[(i, buying)] = max(sell, cooldown)
      return dp[(i, buying)]

    return dfs(0, True)
```

###***7. Coin Change II (Medium)***
https://leetcode.com/problems/coin-change-ii/

```python
from typing import List

class Solution:
  def change(self, amount: int, coins: List[int]) -> int:
    # MEMOIZATION
    # Time: O(n*m)
    # Memory: O(n*m)
    cache = {}
    def dfs(i, a):
      if a == amount:
        return 1
      if a > amount:
        return 0
      if i == len(coins):
        return 0
      if (i, a) in cache:
        return cache[(i, a)]

      cache[(i, a)] = dfs(i, a + coins[i]) + dfs(i + 1, a)
      return cache[(i, a)]

    return dfs(0, 0)

    # DYNAMIC PROGRAMMING
    # Time: O(n*m)
    # Memory: O(n*m)
    dp = [[0] * (len(coins) + 1) for i in range(amount + 1)]
    dp[0] = [1] * (len(coins) + 1)
    for a in range(1, amount + 1):
      for i in range(len(coins) - 1, -1, -1):
        dp[a][i] = dp[a][i + 1]
        if a - coins[i] >= 0:
          dp[a][i] += dp[a - coins[i]][i]
    return dp[amount][0]

    # DYNAMIC PROGRAMMING
    # Time: O(n*m)
    # Memory: O(n) where n = amount
    dp = [0] * (amount + 1)
    dp[0] = 1
    for i in range(len(coins) - 1, -1, -1):
      nextDP = [0] * (amount + 1)
      nextDP[0] = 1

      for a in range(1, amount + 1):
        nextDP[a] = dp[a]
        if a - coins[i] >= 0:
          nextDP[a] += nextDP[a - coins[i]]
      dp = nextDP
    return dp[amount]
```

###***8. Target Sum (Medium)***
https://leetcode.com/problems/target-sum/

```python
from typing import List

class Solution:
  def findTargetSumWays(self, nums: List[int], target: int) -> int:
    dp = {} # (index, total) -> # of ways

    def backtrack(i, total):
      if i == len(nums):
        return 1 if total == target else 0
      if (i, total) in dp:
        return dp[(i, total)]

      dp[(i, total)] = (backtrack(i + 1, total + nums[i]) +
                        backtrack(i + 1, total - nums[i]))
      return dp[(i, total)]
    return backtrack(0, 0)
```

###***9. Interleaving String (Medium)***
https://leetcode.com/problems/interleaving-string/

```python
class Solution:
  def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
    if len(s1) + len(s2) != len(s3): # check against total length
      return False

    # Building 2d grid
    dp = [[False] * (len(s2) + 1) for i in range(len(s1) + 1)]
    dp[len(s1)][len(s2)] = True # initializing conor value as True

    for i in range(len(s1), -1, -1): # Bottom up, bottom right to top left
      for j in range(len(s2), -1, -1):
        if i < len(s1) and s1[i] == s3[i + j] and dp[i + 1][j]:
          dp[i][j] = True
        if j < len(s2) and s2[j] == s3[i + j] and dp[i][j + 1]:
          dp[i][j] = True
    return dp[0][0] # result is in top left
```

###***10. Stone Game (Medium)***
https://leetcode.com/problems/stone-game/

```python
from typing import List

class Solution:
  def stoneGame(self, piles: List[int]) -> bool:
    dp = {} # subarr piles (l, r) -> Max Alice Total

    # Return: Max Alice Total
    def dfs(l, r):
      if l > r:
        return 0
      if (l, r) in dp:
        return dp[(l, r)]

      even = True if (r - l) % 2 else False
      left = piles[l] if even else 0
      right = piles[r] if even else 0

      dp[(l, r)] = max(dfs(l + 1, r) + left, dfs(l, r - 1) + right)
      return dp[(l, r)]

    return dfs(0, len(piles) - 1) > (sum(piles)) // 2
```

###***11. Minimum Path Sum (Medium)***
https://leetcode.com/problems/minimum-path-sum/

```python
from typing import List

class Solution:
  def minPathSum(self, grid: List[List[int]]) -> int:
    ROWS, COLS = len(grid), len(grid[0])

    res = [[float("inf")] * (COLS + 1) for r in range(ROWS + 1)]
    res[ROWS - 1][COLS] = 0 # Set last cell to 0

    for r in range(ROWS - 1, -1, -1):
      for c in range(COLS - 1, -1, -1):
        res[r][c] = grid[r][c] + min(res[r + 1][c], res[r][c + 1]) # current + min of bottom and right
    return res[0][0]
```

###***12. Longest Increasing Path In a Matrix (Hard)***
https://leetcode.com/problems/longest-increasing-path-in-a-matrix/

```python
from typing import List

class Solution:
  def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
    ROWS, COLS = len(matrix), len(matrix[0])
    dp = {}  # (r, c) -> LIP

    def dfs(r, c, prevVal):
      if r < 0 or r == ROWS or c < 0 or c == COLS or matrix[r][c] <= prevVal:
        return 0
      if (r, c) in dp:
        return dp[(r, c)]

      res = 1 # even one cell is considered one path
      res = max(res, 1 + dfs(r + 1, c, matrix[r][c]))
      res = max(res, 1 + dfs(r - 1, c, matrix[r][c]))
      res = max(res, 1 + dfs(r, c + 1, matrix[r][c]))
      res = max(res, 1 + dfs(r, c - 1, matrix[r][c]))
      dp[(r, c)] = res
      return res

    for r in range(ROWS):
      for c in range(COLS):
        dfs(r, c, -1)
    return max(dp.values())
```

###***13. Maximal Square (Medium)***
https://leetcode.com/problems/maximal-square/

```python
from typing import List

class Solution:
  def maximalSquare(self, matrix: List[List[str]]) -> int:
    # Dynamic programming: bottom up
    # Recursive: top down
    ROWS, COLS = len(matrix), len(matrix[0])
    cache = {}  # map each (r, c) -> maxLength of square

    # T: O(m*n) M: O(m*n)
    def helper(r, c):
      if r >= ROWS or c >= COLS:
        return 0

      if (r, c) not in cache:
        # Down, Left, Dia all having "1" will satisfy square
        down = helper(r + 1, c)
        right = helper(r, c + 1)
        diag = helper(r + 1, c + 1)

        cache[(r, c)] = 0
        if matrix[r][c] == "1":
          cache[(r, c)] = 1 + min(down, right, diag) # Can't take max

      return cache[(r, c)] # Already in cache

    helper(0, 0)
    return max(cache.values()) ** 2
```

```python
from typing import List

class Solution:
  def maximalSquare(self, matrix: List[List[str]]) -> int:
    if not matrix:
      return 0

    m, n = len(matrix), len(matrix[0])
    # Initialize a dp table to store the size of the largest square ending at each cell
    dp = [[0] * n for _ in range(m)]
    max_side = 0  # Variable to keep track of the maximum side length of the square

    for i in range(m):
      for j in range(n):
        if matrix[i][j] == '1':
          if i == 0 or j == 0:
            dp[i][j] = 1  # If it's in the first row or column, it can only be itself (either '0' or '1')
          else:
            # Calculate dp[i][j] as the minimum of its top, left, and top-left neighbor + 1
            dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1
          max_side = max(max_side, dp[i][j])  # Update max_side if we find a larger square

    return max_side * max_side  # Return the area of the largest square found
```

###***16. Distinct Subsequences (Hard)***
https://leetcode.com/problems/distinct-subsequences/

```python
class Solution:
  def numDistinct(self, s: str, t: str) -> int:
    cache = {}

    def dfs(i, j): # position of s and t
      if j == len(t):
        return 1
      if i == len(s):
        return 0
      if (i, j) in cache:
        return cache[(i, j)]

      if s[i] == t[j]: # we can decide to chose or not to chose
        cache[(i, j)] = dfs(i + 1, j + 1) + dfs(i + 1, j) # next char is also possible
      else:
        cache[(i, j)] = dfs(i + 1, j)
      return cache[(i, j)]

    return dfs(0, 0)
```

###***17. Edit Distance (Medium)***
https://leetcode.com/problems/edit-distance/

```python
class Solution:
  def minDistance(self, word1: str, word2: str) -> int:
    # Building 2D grid
    dp = [[float("inf")] * (len(word2) + 1) for i in range(len(word1) + 1)]

    # Edge cases
    for j in range(len(word2) + 1):
      dp[len(word1)][j] = len(word2) - j # filling last row
    for i in range(len(word1) + 1):
      dp[i][len(word2)] = len(word1) - i # filling last col

    for i in range(len(word1) - 1, -1, -1): # bottom up
      for j in range(len(word2) - 1, -1, -1):
        if word1[i] == word2[j]: # if equals, i and j will be moved to next
          dp[i][j] = dp[i + 1][j + 1]
        else: # delete (i + 1, j), insert (i, j + 1), replace (i + 1, j + 1)
          dp[i][j] = 1 + min(dp[i + 1][j], dp[i][j + 1], dp[i + 1][j + 1])
    return dp[0][0] # result is in top left
```

###***18. Count Vowels Permutation (Hard)***
https://leetcode.com/problems/count-vowels-permutation/

```python
class Solution:
  def countVowelPermutation(self, n: int) -> int:
    # dp[j][c] = num of strs of len=j,
    # where last char is c
    dp = [[], [1, 1, 1, 1, 1]]

    a, e, i, o, u = 0, 1, 2, 3, 4
    mod = 10**9 + 7

    for j in range(2, n + 1):
      dp.append([0, 0, 0, 0, 0])

      dp[j][a] = (dp[j - 1][e] + dp[j - 1][i] + dp[j - 1][u]) % mod
      dp[j][e] = (dp[j - 1][a] + dp[j - 1][i]) % mod
      dp[j][i] = (dp[j - 1][e] + dp[j - 1][o]) % mod
      dp[j][o] = dp[j - 1][i] # No need to mod here
      dp[j][u] = (dp[j - 1][i] + dp[j - 1][o]) % mod

    return sum(dp[n]) % mod
```

###***19. Burst Balloons (Hard)***
https://leetcode.com/problems/burst-balloons/

```python
from typing import List

class Solution:
  def maxCoins(self, nums: List[int]) -> int:
    nums = [1] + nums + [1] # as per instruction
    dp = {}

    def dfs(l, r):
      if l > r:
        return 0
      if (l, r) in dp:
        return dp[(l, r)]

      dp[(l, r)] = 0
      for i in range(l, r + 1): # inclusive of r, *trick -> burst i at last
        coins = nums[l - 1] * nums[i] * nums[r + 1]
        coins += dfs(l, i - 1) + dfs(i + 1, r) # left and right portions of i
        dp[(l, r)] = max(dp[(l, r)], coins)
      return dp[(l, r)]
    return dfs(1, len(nums) - 2) # exclude [1] from start and end
```

###***21. Regular Expression Matching (Hard)***
https://leetcode.com/problems/regular-expression-matching/

```python
# BOTTOM-UP Dynamic Programming
class Solution:
  def isMatch(self, s: str, p: str) -> bool:
    cache = [[False] * (len(p) + 1) for i in range(len(s) + 1)]
    cache[len(s)][len(p)] = True

    for i in range(len(s), -1, -1):
      for j in range(len(p) - 1, -1, -1):
        match = i < len(s) and (s[i] == p[j] or p[j] == ".")

        if (j + 1) < len(p) and p[j + 1] == "*":
          cache[i][j] = cache[i][j + 2]
          if match:
            cache[i][j] = cache[i + 1][j] or cache[i][j]
        elif match:
          cache[i][j] = cache[i + 1][j + 1]

    return cache[0][0]
```

###***22. Stone Game II (Medium)***
https://leetcode.com/problems/stone-game-ii/

```python
from typing import List

class Solution:
  def stoneGameII(self, piles: List[int]) -> int:
    dp = {}

    # Num of stones alice gets
    def dfs(alice, i, M):
      if i == len(piles):
        return 0
      if (alice, i, M) in dp:
        return dp[(alice, i, M)]

      res = 0 if alice else float("inf")
      total = 0
      for X in range(1, 2 * M + 1):
        if i + X > len(piles):
          break
        total += piles[i + X - 1]
        if alice:
          res = max(res, total + dfs(not alice, i + X, max(M, X)))
        else:
          res = min(res, dfs(not alice, i + X, max(M, X)))
      dp[(alice, i, M)] = res
      return res

    return dfs(True, 0, 1)
```

###***24. Maximum Value of K Coins from Piles (Hard)***
https://leetcode.com/problems/maximum-value-of-k-coins-from-piles/

```python
from typing import List

class Solution:
  def maxValueOfCoins(self, piles: List[List[int]], k: int) -> int:
    n = len(piles)
    dp = [[-1] * (k + 1) for _ in range(n)]

    def dfs(i, coins):
      if i == n:
        return 0
      if dp[i][coins] != -1:
        return dp[i][coins]

      dp[i][coins] = dfs(i + 1, coins) # skip cur pile
      curPile = 0
      for j in range(min(coins, len(piles[i]))):
        curPile += piles[i][j]
        dp[i][coins] = max(dp[i][coins], curPile + dfs(i + 1, coins - j - 1))
      return dp[i][coins]

    return dfs(0, k)
```

###***25. Number of Music Playlists (Hard)***
https://leetcode.com/problems/number-of-music-playlists/

```python
class Solution:
  def numMusicPlaylists(self, n: int, goal: int, k: int) -> int:
    mod = 10**9 + 7
    dp = {}

    def count(cur_goal, old_songs):
      if cur_goal == 0 and old_songs == n:
        return 1
      if cur_goal == 0 or old_songs > n:
        return 0
      if (cur_goal, old_songs) in dp:
        return dp[(cur_goal, old_songs)]

      # Choose new song
      res = (n - old_songs) * count(cur_goal - 1, old_songs + 1)

      if old_songs > k:
        # Choose old song
        res += (old_songs - k) * count(cur_goal - 1, old_songs)
      dp[(cur_goal, old_songs)] = res % mod
      return dp[(cur_goal, old_songs)]

    return count(goal, 0)
```

###***26. Number of Ways to Form a Target String Given a Dictionary (Hard)***
https://leetcode.com/problems/number-of-ways-to-form-a-target-string-given-a-dictionary/

```python
from typing import List

class Solution:
  def numWays(self, words: List[str], target: str) -> int:
    mod = 10**9 + 7

    cnt = defaultdict(int) # (index, char) -> count among all words
    for w in words:
      for i, c in enumerate(w):
        cnt[(i, c)] += 1

    dp = {} # (i, k) => numWays
    # i = index of target, k = index of words[j][k]
    def dfs(i, k):
      if i == len(target):
        return 1
      if k == len(words[0]):
        return 0
      if (i, k) in dp:
        return dp[(i, k)]

      c = target[i]
      dp[(i, k)] = dfs(i, k + 1) # skip k position
      dp[(i, k)] += cnt[(k, c)] * dfs(i + 1, k + 1)
      return dp[(i, k)] % mod

    return dfs(0, 0)
```

###***27. Profitable Schemes (Hard)***
https://leetcode.com/problems/profitable-schemes/

```python
from typing import List

class Solution:
  def profitableSchemes(self, n: int, minProfit: int, group: List[int], profit: List[int]) -> int:
    mod = 1000000007
    dp = [[0] * (n + 1) for _ in range(minProfit + 1)]
    dp[0][0] = 1

    for g, p in zip(group, profit):
      for j in range(n, g - 1, -1):
        dp[minProfit][j] += dp[minProfit][j - g]

      for i in range(max(0, minProfit - p), minProfit):
        for j in range(g, n + 1):
          dp[minProfit][j] += dp[i][j - g]

      for i in range(minProfit - 1, p - 1, -1):
        for j in range(g, n + 1):
          dp[i][j] = (dp[i][j] + dp[i - p][j - g]) % mod

    return sum(dp[minProfit]) % mod
```

###***28. Minimum Cost to Cut a Stick (Hard)***
https://leetcode.com/problems/minimum-cost-to-cut-a-stick/

```python
from typing import List

class Solution:
  def minCost(self, n: int, cuts: List[int]) -> int:
    dp = {}

    def dfs(l, r):
      if r - l == 1: # Base case
        return 0
      if (l, r) in dp:
        return dp[(l, r)]

      res = float("inf")
      for c in cuts:
        if l < c < r:
          res = min(res, (r - l) + dfs(l, c) + dfs(c, r))

      # If stick can't be cut, we want to return 0
      # Adding to dp cache
      dp[(l, r)] = res = 0 if res == float("inf") else res
      return res

    return dfs(0, n)
```

```python
from typing import List

class Solution:
  def minCost(self, n: int, cuts: List[int]) -> int:
    cuts.sort()
    cuts = [0] + cuts + [n]
    m = len(cuts)

    # Initialize DP table with large values
    dp = [[float('inf')] * m for _ in range(m)]

    # Base case: adjacent cuts cost 0
    for i in range(m - 1):
      dp[i][i + 1] = 0

    # Fill DP table
    for length in range(2, m):
      for i in range(m - length):
        j = i + length
        for k in range(i + 1, j):
          dp[i][j] = min(dp[i][j], dp[i][k] + dp[k][j] + cuts[j] - cuts[i])

    return dp[0][m - 1]
```
