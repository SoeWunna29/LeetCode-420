#10. Backtracking (18 Questions)

###***1. Subsets (Medium)***
https://leetcode.com/problems/subsets/

```python
from typing import List

class Solution:
  def subsets(self, nums: List[int]) -> List[List[int]]:
    res = []

    subset = []
    def dfs(i): # i is the index here
      if i >= len(nums):
        res.append(subset.copy())
        return

      # decision to include nums[i]
      subset.append(nums[i])
      dfs(i + 1)

      # decision NOT to include nums[i]
      subset.pop()
      dfs(i + 1)

    dfs(0)
    return res
```

###***2. Combination Sum (Medium)***
https://leetcode.com/problems/combination-sum/

```python
from typing import List

class Solution:
  def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
    res = []

    def dfs(i, cur, total):
      if total == target:
        res.append(cur.copy())
        return
      if i >= len(candidates) or total > target:
        return

      cur.append(candidates[i])
      dfs(i, cur, total + candidates[i]) # include i
      cur.pop() # cleaning before going down to other decision
      dfs(i + 1, cur, total) # not include i

    dfs(0, [], 0)
    return res
```

###***3. Combinations (Medium)***
https://leetcode.com/problems/combinations/

```python
from typing import List

class Solution:
  def combine(self, n: int, k: int) -> List[List[int]]:
    res = []

    def backtrack(start, comb):
      # Base case
      if len(comb) == k:
        res.append(comb.copy()) # Adding copy to that it won't be updated with recursive call
        return

      # Main algorithm
      for i in range(start, n + 1):
        comb.append(i)
        backtrack(i + 1, comb) # i + 1, only add greater val
        comb.pop() # Cleaning the val before next iteration

    backtrack(1, [])
    return res
```

###***4. Permutations (Medium)***
https://leetcode.com/problems/permutations/

```python
from typing import List

class Solution:
  def permute(self, nums: List[int]) -> List[List[int]]:
    result = []

    # base case
    if (len(nums) == 1):
      return [nums.copy()] # nums[:] can also be used

    for i in range(len(nums)):
      n = nums.pop(0)
      perms = self.permute(nums)

      for perm in perms:
        perm.append(n)
      result.extend(perms)
      nums.append(n)

    return result
```

###***5. Subsets II (Medium)***
https://leetcode.com/problems/subsets-ii/

```python
from typing import List

class Solution:
  def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
    res = []
    nums.sort()

    def backtrack(i, subset):
      if i == len(nums):
        res.append(subset[::]) # subset.copy()
        return

      # All subsets that include nums[i]
      subset.append(nums[i])
      backtrack(i + 1, subset)
      subset.pop()

      # All subsets that don't include nums[i]
      # Duplicates will be next to each other since array is sorted
      while i + 1 < len(nums) and nums[i] == nums[i + 1]:
        i += 1
      backtrack(i + 1, subset)

    backtrack(0, [])
    return res
```

###***6. Combination Sum II (Medium)***
https://leetcode.com/problems/combination-sum-ii/

```python
from typing import List

class Solution:
  def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
    candidates.sort() # to make skipping duplicates easy
    res = []

    def backtrack(cur, pos, target):
      if target == 0: # we found the combination
        res.append(cur.copy()) # cur is just reference, add copy() of it
        return
      if target <= 0:
        return

      prev = -1 # default value
      for i in range(pos, len(candidates)): # starting pos to end of candidates
        if candidates[i] == prev: # is equal to current candidate, skip
          continue
        cur.append(candidates[i]) # prework before calling recursive func
        backtrack(cur, i + 1, target - candidates[i])
        cur.pop()
        prev = candidates[i]

    backtrack([], 0, target)
    return res
```

###***7. Permutations II (Medium)***
https://leetcode.com/problems/permutations-ii/

```python
from typing import List

class Solution:
  def permuteUnique(self, nums: List[int]) -> List[List[int]]:
    res = []
    perm = []
    # Eiminate duplicates by transforming list into count hash map
    count = {n:0 for n in nums}
    for n in nums:
      count[n] += 1

    def dfs():
      # Base case
      if len(perm) == len(nums):
        res.append(perm.copy()) # We are adding copy since original val will be updated
        return

      for n in count:
        if count[n] > 0: # Enough val?
          perm.append(n)
          count[n] -= 1

          dfs() # Recursive call

          # Cleaning up
          count[n] += 1
          perm.pop()

    dfs()
    return res
```

###***8. Word Search (Medium)***
https://leetcode.com/problems/word-search/

```python
from typing import List

class Solution:
  def exist(self, board: List[List[str]], word: str) -> bool:
    ROWS, COLS = len(board), len(board[0])
    path = set()

    def dfs(r, c, i):
      if i == len(word):
        return True
      if (r < 0 or c < 0 or
          r >= ROWS or c >= COLS or
          word[i] != board[r][c] or
          (r, c) in path):
          return False

      path.add((r, c))
      res = (dfs(r + 1, c, i + 1) or
             dfs(r - 1, c, i + 1) or
             dfs(r, c + 1, i + 1) or
             dfs(r, c - 1, i + 1))
      path.remove((r, c))
      return res

    for r in range(ROWS):
      for c in range(COLS):
        if dfs(r, c, 0):
          return True
    return False
```

###***9. Palindrome Partitioning (Medium)***
https://leetcode.com/problems/palindrome-partitioning/

```python
from typing import List

class Solution:
  def partition(self, s: str) -> List[List[str]]:
    res, part = [], []

    def dfs(i): # index
      if i >= len(s): # reached the end of the string
        res.append(part.copy())
        return
      for j in range(i, len(s)):
        if self.isPali(s, i, j):
            part.append(s[i:j+1])
            dfs(j + 1)
            part.pop()

    dfs(0) # start from first index
    return res

  def isPali(self, s, l, r):
    while l < r:
      if s[l] != s[r]:
        return False
      l, r = l + 1, r - 1
    return True
```

###***10. Restore IP Addresses (Medium)***
https://leetcode.com/problems/restore-ip-addresses/

```python
from typing import List

class Solution:
  def restoreIpAddresses(self, s: str) -> List[str]:
    res = []

    if len(s) > 12: # Base case
      return res

    def backtrack(i, dots, curIP):
      if dots == 4 and i == len(s):
        res.append(curIP[:-1]) # remove last dot "."
        return

      if dots > 4: # Not valid IP
        return

      for j in range(i, min(i + 3, len(s))): # min(i + 3, len(s)), either end of string or i + 3
        if int(s[i:j+1]) < 256 and (i == j or s[i] != "0"): # start i, end j +  1 (not inclusive), length 1 or no leading zero
          backtrack(j + 1, dots + 1, curIP + s[i:j+1] + ".")

    backtrack(0, 0, "")
    return res
```

###***11. Letter Combinations of a Phone Number (Medium)***
https://leetcode.com/problems/letter-combinations-of-a-phone-number/

```python
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
      res = []
      digitToChar = {
          "2": "abc",
          "3": "def",
          "4": "ghi",
          "5": "jkl",
          "6": "mno",
          "7": "qprs",
          "8": "tuv",
          "9": "wxyz",
      }

      def backtrack(i, curStr):
        if len(curStr) == len(digits): # i >= len(digits), base case
          res.append(curStr)
          return
        for c in digitToChar[digits[i]]:
          backtrack(i + 1, curStr + c) # eg: i -> "23", curStr = "ad"

      if digits: # edge case
        backtrack(0, "")

      return res
```

###***12. Matchsticks to Square (Medium)***
https://leetcode.com/problems/matchsticks-to-square/

```python
from typing import List

class Solution:
  def makesquare(self, matchsticks: List[int]) -> bool:
    length = sum(matchsticks) // 4
    sides = [0] * 4 # L T R B

    if sum(matchsticks) / 4 != length:
      return False

    matchsticks.sort(reverse=True) # To check largest matchstick > length, False

    def backtrack(i):
      if i == len(matchsticks):
        return True

      for j in range(4): # Go through 4 sides
        if sides[j] + matchsticks[i] <= length: # cur side <= length of each side
          sides[j] += matchsticks[i]
          if backtrack(i + 1):
            return True
          sides[j] -= matchsticks[i]
      return False

    return backtrack(0)
```

###***13. Splitting a String Into Descending Consecutive Values (Medium)***
https://leetcode.com/problems/splitting-a-string-into-descending-consecutive-values/

```python
class Solution:
  def splitString(self, s: str) -> bool:
    def dfs(index, prev):
      if index == len(s):
        return True

      for j in range(index, len(s)):
        val = int(s[index:j + 1])
        if val + 1 == prev and dfs(j + 1, val):
          return True
      return False

    for i in range(len(s) - 1):
      val = int(s[:i + 1])
      if dfs(i + 1, val): return True

    return False
```

###***14. Find Unique Binary String (Medium)***
https://leetcode.com/problems/find-unique-binary-string/

```python
from typing import List

class Solution:
  def findDifferentBinaryString(self, nums: List[str]) -> str:
    strSet = { s for s in nums }

    def backtrack(i, cur):
      if i == len(nums): # Base case
        res = "".join(cur)
        return None if res in strSet else res # Return the res if it's not in nums

      res = backtrack(i+1, cur)
      if res: return res

      cur[i] = "1" # Second decision
      res = backtrack(i+1, cur)
      if res: return res

    return backtrack(0, ["0" for s in nums]) # First decision, "0"
```

###***15. Maximum Length of a Concatenated String With Unique Characters (Medium)***
https://leetcode.com/problems/maximum-length-of-a-concatenated-string-with-unique-characters/

```python
from typing import List

class Solution:
  def maxLength(self, arr: List[str]) -> int:
    charSet = set()

    def overlap(charSet, s):
      c = Counter(charSet) + Counter(s)
      return max(c.values()) > 1 # Overlap if max c.val > 1

    def backtrack(i):
      if i == len(arr): # Base case
        return len(charSet)
      res = 0

      if not overlap(charSet, arr[i]): # Include
        for c in arr[i]:
          charSet.add(c)
        res = backtrack(i + 1)
        for c in arr[i]:
          charSet.remove(c)

      # Not include
      return max(res, backtrack(i + 1)) # don't concatenate arr[i]

    return backtrack(0)
```

###***16. Partition to K Equal Sum Subsets (Medium)***
https://leetcode.com/problems/partition-to-k-equal-sum-subsets/

```python
from typing import List

class Solution:
  def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
    if sum(nums) % k:
      return False
    nums.sort(reverse=True)
    target = sum(nums) / k
    used = [False] * len(nums)

    def backtrack(i, k, subsetSum): # k = remaining subsets to build
      if k == 0: # Base Cases
        return True
      if subsetSum == target: # Found the subset
        return backtrack(0, k - 1, 0) # Building a new subset

      for j in range(i, len(nums)):
        if used[j] or subsetSum + nums[j] > target or (j > 0 and nums[j] == nums[j-1] and not used[j-1]): # Don't need to backtrack / Edited
          continue

        used[j] = True
        if backtrack(j + 1, k, subsetSum + nums[j]):
          return True
        used[j] = False # Cleaning up after backtracking

      return False

    return backtrack(0, k, 0)
```

###***17. N Queens (Hard)***
https://leetcode.com/problems/n-queens/

```python
from typing import List

class Solution:
  def solveNQueens(self, n: int) -> List[List[str]]:
    col = set()
    posDiag = set() # (r + c) positive diagonal
    negDiag = set() # (r - c) negative diagonal

    res = []
    board = [["."] * n for i in range(n)]

    def backtrack(r):
      if r == n: # base case
        copy = ["".join(row) for row in board]
        res.append(copy)
        return

      for c in range(n):
        if c in col or (r + c) in posDiag or (r - c) in negDiag:
          continue

        col.add(c)
        posDiag.add(r + c)
        negDiag.add(r - c)
        board[r][c] = "Q"

        backtrack(r + 1)

        col.remove(c)
        posDiag.remove(r + c)
        negDiag.remove(r - c)
        board[r][c] = "."

    backtrack(0)
    return res
```

###***18. N Queens II (Hard)***
https://leetcode.com/problems/n-queens-ii/

```python
class Solution:
  def totalNQueens(self, n: int) -> int:
    col = set()
    posDiag = set() # (r + c), Bottom Left
    negDiag = set() # (r - c), Bottom Right

    res = 0
    def backtrack(r):
      nonlocal res
      if r == n: # Base case
        res += 1
        return

      for c in range(n):
        if c in col or (r + c) in posDiag or (r - c) in negDiag: # We can't put a queen
          continue

        col.add(c)
        posDiag.add(r + c)
        negDiag.add(r - c)

        backtrack(r + 1)

        col.remove(c)
        posDiag.remove(r + c)
        negDiag.remove(r - c)

    backtrack(0)
    return res
```
