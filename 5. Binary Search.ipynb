###***1. Binary Search (Easy)***
https://leetcode.com/problems/binary-search/

```python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) -1

        while l <= r:
            m = l + ((r - l) // 2)
            if nums[m] < target:
                l = m + 1
            elif nums[m] > target:
                r = m - 1
            else:
                return m
        return -1
```

###***2. Search Insert Position (Easy)***
https://leetcode.com/problems/search-insert-position/

```python
from typing import List

class Solution:
  def searchInsert(self, nums: List[int], target: int) -> int:
    # O(log n) and O(1)
    l, r = 0, len(nums) - 1

    while l <= r:
      mid = (l + r) // 2
      if target == nums[mid]:
        return mid
      if target > nums[mid]:
        l = mid + 1
      else:
        r = mid - 1

    # l represents where the target should be inserted to maintain the sorted order.
    return l # loop terminates when l > r, and the target is not found
```

```python
class Solution:
  def guessNumber(self, n: int) -> int:
    # return a num btw 1,..,n
    low = 1
    high = n

    while True:
      mid = low + (high - low) // 2
      myGuess = guess(mid)
      if myGuess == 1:
        low = mid + 1
      elif myGuess == -1:
        high = mid - 1
      else:
        return mid
```

###***4. Arranging Coins (Easy)***
https://leetcode.com/problems/arranging-coins/

```python
class Solution:
  def arrangeCoins(self, n: int) -> int:
    l, r = 1, n
    res = 0
    while l <=r:
      mid = (l+r)//2
      coins = (mid /2) * (mid+1)
      if coins > n:
        r = mid - 1
      else:
        l = mid + 1
        res = max(mid, res)
    return res
```

###***5. Squares of a Sorted Array (Easy)***
https://leetcode.com/problems/squares-of-a-sorted-array/

```python
from typing import List

class Solution:
  def sortedSquares(self, nums: List[int]) -> List[int]:
    res = []
    l, r = 0, len(nums) - 1
    while l <= r:
      if nums[l] ** 2 > nums[r] ** 2:
        res.append(nums[l] ** 2)
        l += 1
      else:
        res.append(nums[r] ** 2)
        r -= 1

    return res[::-1]
```

###***6. Valid Perfect Square (Easy)***
https://leetcode.com/problems/valid-perfect-square/

```python
class Solution:
  def isPerfectSquare(self, num: int) -> bool:
    l, r = 1, num
    while l <= r:
      mid = (l + r) // 2
      if mid * mid > num: # look for smaller mid
        r = mid - 1
      elif mid * mid < num: # look for larger mid
        l = mid + 1
      else:
        return True
    return False
```

###***7. Sqrt(x) (Easy)***

```python
class Solution:
  def mySqrt(self, x: int) -> int:
    l, r = 0, x
    res = 0

    while l <= r:
      m = l + ((r - l) // 2)
      if m**2 > x:
        r = m - 1
      elif m**2 < x:
        l = m + 1
        res = m # greatest mid val that is less than x
      else: # m**2 == x
        return m
    return res
```

###***8. Single Element in a Sorted Array (Medium)***
https://leetcode.com/problems/single-element-in-a-sorted-array/

```python
from typing import List

class Solution:
  def singleNonDuplicate(self, nums: List[int]) -> int:
    l, r = 0, len(nums) - 1

    while l <= r:
      m = l + ((r - l) // 2)
      # Left is out of bound or not equal and right is out bound or not equal
      if ((m - 1 < 0 or nums[m - 1] != nums[m]) and
          (m + 1 == len(nums) or nums[m] != nums[m + 1])):
        return nums[m]

      leftSize = m - 1 if nums[m - 1] == nums[m] else m
      if leftSize % 2: # Odd num val on left side
        r = m - 1
      else: # Right side
        l = m + 1
```

###***9. Capacity to Ship Packages (Medium)***
https://leetcode.com/problems/capacity-to-ship-packages-within-d-days/

```python
class Solution:
  def shipWithinDays(self, weights: List[int], days: int) -> int:
    l, r = max(weights), sum(weights)
    min_cap = r

    def canShip(cap):
      ships, curCap = 1, cap
      for w in weights:
        if curCap - w < 0:
          ships += 1
          curCap = cap
        curCap -= w
      return ships <= days

    while l <= r:
      cap = (l + r) // 2
      if canShip(cap):
        min_cap = min(min_cap, cap)
        r = cap - 1
      else:
        l = cap + 1

    return min_cap
```

###***10. Find Peak Element (Medium)***
https://leetcode.com/problems/find-peak-element/

```python
from typing import List

class Solution:
  def findPeakElement(self, nums: List[int]) -> int:
    left, right = 0, len(nums) - 1
    while left < right:
      mid = left + (right - left) // 2
      if nums[mid] > nums[mid + 1]: # Left neighbor is greater
        right = mid
      else: # Right neighbor is greater
        left = mid + 1
    return left
```

###***11. Successful Pairs of Spells and Potions (Medium)***
https://leetcode.com/problems/successful-pairs-of-spells-and-potions/

```python
from typing import List

class Solution:
  def successfulPairs(self, spells: List[int], potions: List[int], success: int) -> List[int]:
    # O(nlogm + mlogm)
    # Sort potions and Binary search

    potions.sort()
    res = []

    for s in spells:
      l, r = 0, len(potions) - 1
      idx = len(potions) # Find weakest potion (furthest left) that works

      while l <= r:
        m = (l + r) // 2
        if s * potions[m] >= success:
          r = m - 1
          idx = m
        else:
          l = m + 1

      res.append(len(potions) - idx) # In case we couldn't update idx = m, it will be zero
    return res
```

###***12. Search a 2D Matrix (Medium)***
https://leetcode.com/problems/search-a-2d-matrix/

```python
from typing import List

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
      ROWS, COLS = len(matrix), len(matrix[0])
      # Binary search for row in matrix that might contain target val
      top, bot = 0, ROWS - 1
      while top <= bot:
        row = (top + bot) // 2 # finding mid row
        if target > matrix[row][-1]:
          top = row + 1
        elif target < matrix[row][0]:
          bot = row - 1
        else:
          break

      if not (top <= bot): # none of the rows contain target val
        return False

      # Binary search for target val in the row
      row = (top + bot) // 2
      l, r = 0, COLS - 1
      while l <= r:
        m = (l + r) // 2
        if target > matrix[row][m]:
          l = m + 1
        elif target < matrix[row][m]:
          r = m - 1
        else:
          return True
      return False
```

###***14. Minimize the Maximum Difference of Pairs (Medium)***
https://leetcode.com/problems/minimize-the-maximum-difference-of-pairs/

```python
from typing import List

class Solution:
  def minimizeMax(self, nums: List[int], p: int) -> int:
    def isValid(threshold):
      i, cnt = 0, 0
      while i < len(nums) - 1:
        if abs(nums[i] - nums[i + 1]) <= threshold:
          cnt += 1
          i += 2
        else:
          i += 1
        if cnt == p:
          return True
      return False

    if p == 0: return 0

    nums.sort()
    l, r = 0, 10**9
    res = 10**9

    while l <= r:
      m = l + (r - l) // 2
      if isValid(m):
        res = m
        r = m - 1
      else:
        l = m + 1
    return res
```

###***15. Find Minimum In Rotated Sorted Array (Medium)***
https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/

```python
# Left, Right & Middle Pointer Problem

def find_min(nums):
    result = nums[0]
    left = 0
    right = len(nums) - 1

    while left <= right:
        if nums[left] < nums[right]:
            result = min(result, nums[left])
            break

        middle = (left + right) // 2
        result = min(result, nums[middle])
        if nums[middle] >= nums[left]:
            left = middle + 1
        else:
            right = middle - 1

    return result

nums = [4, 5, 6, 7, 0, 1, 2]
print(find_min(nums))
```

###***16. Search In Rotated Sorted Array (Medium)***
https://leetcode.com/problems/search-in-rotated-sorted-array/

```python
# Left, Right & Middle Pointer Problem
# Binary Search Problem

def find_num(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        middle = (left + right) // 2
        if target == nums[middle]:
            return middle

        # left sorted portion
        if nums[left] <= nums[middle]:
            if target > nums[middle] or target < nums[left]: # t < leftmost
                left = middle + 1
            else:
                right = middle - 1

        # right sorted portion
        else:
            if target < nums[middle] or target > nums[right]: # t > rightmost
                right = middle - 1
            else:
                left = middle + 1

    return -1

nums = [4, 5, 6, 7, 0, 1, 2]
target = 2
print(find_num(nums, target))
```

###***17. Search In Rotated Sorted Array II (Medium)***
https://leetcode.com/problems/search-in-rotated-sorted-array-ii/

```python
from typing import List

class Solution:
  def search(self, nums: List[int], target: int) -> bool:
    l, r = 0, len(nums) - 1
    while l <= r:
      m = l + (r - l) // 2
      if nums[m] == target: return True

      if nums[l] < nums[m]: # Left portion
        if nums[l] <= target < nums[m]:
          r = m - 1
        else:
          l = m + 1
      elif nums[l] > nums[m]: # Right portion
        if nums[m] < target <= nums[r]:
          l = m + 1
        else:
          r = m - 1
      else:
        l += 1
    return False
```

###***18. Time Based Key Value Store (Medium)***
https://leetcode.com/problems/time-based-key-value-store/

```python
class TimeMap:

  def __init__(self):
    self.store = {} # key: list of [val, timestamp]

  def set(self, key: str, value: str, timestamp: int) -> None:
    if key not in self.store:
      self.store[key] = []
    self.store[key].append([value, timestamp])

  def get(self, key: str, timestamp: int) -> str:
    res = ""
    values = self.store.get(key, [])

    # binary search
    l, r = 0, len(values) - 1
    while l <= r:
      m = (l + r) // 2
      if values[m][1] <= timestamp:
        res = values[m][0]
        l = m + 1
      else:
        r = m - 1
    return res
```

###***19. Find First And Last Position of Element In Sorted Array (Medium)***
https://leetcode.com/problems/find-first-and-last-position-of-element-in-sorted-array/

```python
from typing import List

class Solution:
  def searchRange(self, nums: List[int], target: int) -> List[int]:
    left = self.binSearch(nums, target, True)
    right = self.binSearch(nums, target, False)
    return [left, right]

  # leftBias=[True/False], if false, res is rightBiased
  def binSearch(self, nums, target, leftBias):
    l, r = 0, len(nums) - 1
    i = -1
    while l <= r:
      m = (l + r) // 2
      if target > nums[m]:
        l = m + 1
      elif target < nums[m]:
        r = m - 1
      else:
        i = m
        if leftBias:
          r = m - 1
        else:
          l = m + 1
    return i
```

###***20. Maximum Number of Removable Characters (Medium)***
https://leetcode.com/problems/maximum-number-of-removable-characters/

```python
class Solution:
  def maximumRemovals(self, s: str, p: str, removable: List[int]) -> int:
    def isSubsequence(s, p, k):
      removed_indices = set(removable[:k])
      i, j = 0, 0
      while i < len(s) and j < len(p):
        if i in removed_indices or s[i] != p[j]:
          i += 1
        else:
          i += 1
          j += 1
      return j == len(p)

    left, right = 0, len(removable)
    result = 0
    while left <= right:
      mid = (left + right) // 2
      if isSubsequence(s, p, mid):
        result = mid
        left = mid + 1
      else:
        right = mid - 1
    return result
```

###***21. Populating Next Right Pointers In Each Node (Medium)***
https://leetcode.com/problems/populating-next-right-pointers-in-each-node/

```python
class Node:
  def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
    self.val = val
    self.left = left
    self.right = right
    self.next = next

class Solution:
  def connect(self, node):
    cur, nxt = node, node.left if node else None
    # BFS
    while cur and nxt:
      cur.left.next = cur.right
      if cur.next:
        cur.right.next = cur.next.left

      cur = cur.next
      if not cur:
        cur = nxt
        nxt = cur.left
    return node
```

###***22. Search Suggestions System (Medium)***
https://neetcode.io/practice#:~:text=Search%20Suggestions%20System

```python
from typing import List

class Solution:
  def suggestedProducts(self, products: List[str], searchWord: str) -> List[List[str]]:
    res = []
    products.sort()

    l, r = 0, len(products) - 1
    for i in range(len(searchWord)):
      c = searchWord[i]

      while l <= r and (len(products[l]) <= i or products[l][i] != c):
        l += 1
      while l <= r and (len(products[r]) <= i or products[r][i] != c):
        r -= 1

      res.append([])
      remain = r - l + 1
      for j in range(min(3, remain)): # We only want at most 3
        res[-1].append(products[l + j])
    return res
```

###***23. Split Array Largest Sum (Hard)***
https://leetcode.com/problems/split-array-largest-sum/

```python
from typing import List

class Solution:
  def splitArray(self, nums: List[int], m: int) -> int:
    def canSplit(largest):
      subarray = 0
      curSum = 0
      for n in nums:
        curSum += n
        if curSum > largest:
          subarray += 1
          curSum = n
      return subarray + 1 <= m

    l, r = max(nums), sum(nums)
    res = r
    while l <= r:
      mid = l + ((r - l) // 2)
      if canSplit(mid):
        res = mid
        r = mid - 1
      else:
        l = mid + 1
    return res
```

###***24. Median of Two Sorted Arrays (Hard)***
https://leetcode.com/problems/median-of-two-sorted-arrays/

```python
class Solution:
  def findMedianSortedArrays(self, nums1, nums2):
    # Combine and sort the arrays
    nums = sorted(nums1 + nums2)
    mid = len(nums) // 2

    # Check if the length of the combined array is odd
    if len(nums) % 2 != 0:
      return float(nums[mid])
    else:
        # For even length, the median is the average of the two middle numbers
      return (nums[mid - 1] + nums[mid]) / 2.0
```

```python
class Solution:
  def findMedianSortedArrays(self, nums1, nums2):
    A, B = nums1, nums2
    total = len(nums1) + len(nums2)
    half = total // 2

    if len(B) < len(A):
      A, B = B, A
    # log(min(n, m))
    l, r = 0, len(A) - 1
    while True:
      i = (l + r) // 2 # A
      j = half - i - 2 # B

      Aleft = A[i] if i >= 0 else float("-infinity")
      Aright = A[i + 1] if (i + 1) < len(A) else float("infinity")
      Bleft = B[j] if j >= 0 else float("-infinity")
      Bright = B[j + 1] if (j + 1) < len(B) else float("infinity")

      # partition is correct
      if Aleft <= Bright and Bleft <= Aright:
        # odd
        if total % 2:
          return min(Aright, Bright)
        # even
        return (max(Aleft, Bleft) + min(Aright, Bright)) / 2
      elif Aleft > Bright:
        r = i - 1
      else:
        l = i + 1
```
