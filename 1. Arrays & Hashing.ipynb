###***1. Contains Duplicate (Easy)***
https://leetcode.com/problems/contains-duplicate/

```python
def contains_duplicate(nums):
    hash_set = set()

    for n in nums:
        if n in hash_set:
            return True
        hash_set.add(n)
    return False

print(contains_duplicate([1, 2, 3, 5, 5, 1, 2]))

# set consumes less space than list
```

###***2. Valid Anagram (Easy)***
https://leetcode.com/problems/valid-anagram/

```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False

    s_freq = {}
    t_freq = {}

    # Count character frequencies in string s
    for char in s:
        s_freq[char] = s_freq.get(char, 0) + 1

    # Count character frequencies in string t
    for char in t:
        t_freq[char] = t_freq.get(char, 0) + 1

    return s_freq == t_freq
```

###***4. Replace Elements With Greatest Element On Right Side (Easy)***
https://leetcode.com/problems/replace-elements-with-greatest-element-on-right-side/

```python
from typing import List

class Solution:
  def replaceElements(self, arr: List[int]) -> List[int]:
    rightMax = -1 # Initial max -1
    for i in range(len(arr) - 1, -1, -1): # Reverse iteration
      newMax = max(rightMax, arr[i]) # new max = max(old max, arr[i])
      arr[i] = rightMax
      rightMax = newMax
    return arr
```

```python
class Solution:
  def isSubsequence(self, s: str, t: str) -> bool:
    i, j = 0, 0 # i for s, j for t

    while i < len(s) and j < len(t):
      if s[i] == t[j]:
        i += 1
      j += 1

    return True if i == len(s) else False
```

###***7. Two Sum (Easy)***
https://leetcode.com/problems/two-sum/

```python
from typing import List

def two_sum(nums: List[int], target: int) -> List[int]:
    prev_map = {}  # val: index

    for key, value in enumerate(nums):
        difference = target - value
        if difference in prev_map:
            return [prev_map[difference], key]
        prev_map[value] = key
    return prev_map

print(two_sum([1, 3, 5, 7], 10))
```

###***8. Longest Common Prefix (Easy)***
https://leetcode.com/problems/longest-common-prefix/

```python
from typing import List

class Solution:
  def longestCommonPrefix(self, strs: List[str]) -> str:
    res = ""

    for i in range(len(strs[0])): # check against first string in the list
      for s in strs:
        if i == len(s) or s[i] != strs[0][i]: # i == len(s) out of bound
          return res
      res += strs[0][i]

    return res
```

###***9. Group Anagrams (Medium)***
https://leetcode.com/problems/group-anagrams/

```python
class Solution:
  def group_anagram(self, strs):
    anagrams = {}

    for word in strs:
        sorted_word = ''.join(sorted(word))

        if sorted_word in anagrams:
            anagrams[sorted_word].append(word)
        else:
            anagrams[sorted_word] = [word]

    return list(anagrams.values())

sol = Solution()
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(sol.group_anagram(strs))
```

###***10. Pascals Triangle (Easy)***
https://leetcode.com/problems/pascals-triangle/

```python
from typing import List

class Solution:
  def generate(self, numRows: int) -> List[List[int]]:
    res = [[1]] # initialization

    for i in range(numRows - 1): # we already built one row
      temp = [0] + res[-1] + [0] # previous row surrounded by [0]
      row = [] # current row
      for j in range(len(res[-1]) + 1):
        row.append(temp[j] + temp[j + 1]) # two pointers on previous row
      res.append(row)
    return res
```

###***11. Remove Element (Easy)***
https://leetcode.com/problems/remove-element/

```python
class Solution:
  def removeElement(self, nums: List[int], val: int) -> int:
    # Avoid unnecessary copy operations in a previous solution, when k == i and nums[i] != val
    # by swapping nums[i] and the last element of the array (nums[n])
    n = len(nums)
    i = 0

    while i < n:
      if nums[i] == val:
        nums[i], nums[n - 1] = nums[n - 1], nums[i] # swap current and last element
        n -= 1  # decrement the length of the array by discarding the last element
      else:
        i += 1
    return n
```

###***13. Isomorphic Strings (Easy)***
https://leetcode.com/problems/isomorphic-strings/

```python
class Solution:
  def isIsomorphic(self, s: str, t: str) -> bool:
    mapST, mapTS = {}, {} # Use two hash maps, S to T and T to S

    for i in range(len(s)):
      c1, c2 = s[i], t[i]
      if (c1 in mapST and mapST[c1] != c2) or (c2 in mapTS and mapTS[c2] != c1): # Can't have different mapping values
        return False
      mapST[c1] = c2
      mapTS[c2] = c1

    return True
```

###***14. Can Place Flowers (Easy)***
https://leetcode.com/problems/can-place-flowers/

```python
from typing import List

class Solution:
  def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
    # Solution with O(1) space complexity
    empty = 0 if flowerbed[0] else 1

    for f in flowerbed:
      if f:
        n -= int((empty - 1) / 2)  # int division, round toward zero
        empty = 0
      else:
        empty += 1

    n -= (empty) // 2
    return n <= 0
```

###***17. Find Pivot Index (Easy)***
https://leetcode.com/problems/find-pivot-index/

```python
class Solution:
  def pivotIndex(self, nums: List[int]) -> int:
    total = sum(nums)  # O(n)

    leftSum = 0
    for i in range(len(nums)):
      rightSum = total - nums[i] - leftSum # left - i - right
      if leftSum == rightSum:
        return i
      leftSum += nums[i]
    return -1
```

###***18. Range Sum Query - Immutable (Easy)***

```python
class NumArray:
  def __init__(self, nums: List[int]):
    self.prefix = []
    cur = 0
    for n in nums: # Calculate the prefix sum
      cur += n
      self.prefix.append(cur)

  def sumRange(self, left: int, right: int) -> int:
    rightSum = self.prefix[right]
    leftSum = self.prefix[left - 1] if left > 0 else 0
    return rightSum - leftSum # this is key idea, sumRange = r - l
```

###***21. Word Pattern (Easy)***
https://leetcode.com/problems/word-pattern/

```python
class Solution:
  def wordPattern(self, pattern: str, s: str) -> bool:
    words = s.split(" ")
    if len(pattern) != len(words):
      return False

    charToWord = {}
    wordToChar = {}

    for c, w in zip(pattern, words): # iterate through 2 loops
      if c in charToWord and charToWord[c] != w: # c is pattern, w in word
        return False
      if w in wordToChar and wordToChar[w] != c:
        return False
      charToWord[c] = w
      wordToChar[w] = c
    return True
```

###***23. Design HashMap (Easy)***
https://leetcode.com/problems/design-hashmap/

```python
class ListNode:
  def __init__(self, key=-1, val=-1, next=None):
    self.key = key
    self.val = val
    self.next = next

class MyHashMap:
  def __init__(self):
    self.map = [ListNode() for i in range(1000)]

  def hashcode(self, key):
    return key % len(self.map)

  def put(self, key: int, value: int) -> None:
    cur = self.map[self.hashcode(key)]
    while cur.next:
      if cur.next.key == key:
        cur.next.val = value
        return
      cur = cur.next
    cur.next = ListNode(key, value)

  def get(self, key: int) -> int:
    cur = self.map[self.hashcode(key)].next
    while cur and cur.key != key:
      cur = cur.next
    if cur:
      return cur.val
    return -1

  def remove(self, key: int) -> None:
    cur = self.map[self.hashcode(key)]
    while cur.next and cur.next.key != key:
      cur = cur.next
    if cur and cur.next:
      cur.next = cur.next.next
```

###***24. Sort an Array (Medium)***
https://leetcode.com/problems/sort-an-array/

```python
from typing import List

class Solution:
  def sortArray(self, nums: List[int]) -> List[int]: # Merge Sort
    def merge(arr, L, M, R):
      left, right = arr[L:M+1], arr[M+1:R+1]
      i, j, k = L, 0, 0  # j points left arr, k points right arr

      while j < len(left) and k < len(right):
        if left[j] <= right[k]:
          arr[i] = left[j]
          j += 1
        else:
          arr[i] = right[k]
          k += 1
        i += 1  # i will be incremented anyways

      while j < len(left): # remaining values
        arr[i] = left[j]
        j += 1
        i += 1
      while k < len(right):
        arr[i] = right[k]
        k += 1
        i += 1

    def mergeSort(arr, l, r):
      if l == r:
        return arr

      m = (l + r) // 2
      mergeSort(arr, l, m)
      mergeSort(arr, m + 1, r)
      merge(arr, l, m, r)
      return arr

    return mergeSort(nums, 0, len(nums) - 1)

# Quick Sort and Merge Sort (Most important)
```


###***25. Top K Frequent Elements (Medium)***
https://leetcode.com/problems/top-k-frequent-elements/

```python
from collections import Counter
import heapq

def topKFrequent(nums, k):
  # Count the frequency of each element
  counts = Counter(nums)

  # Use a min heap to keep track of the k most frequent elements
  heap = []
  for num, freq in counts.items():
    heapq.heappush(heap, (freq, num))
    if len(heap) > k:
        heapq.heappop(heap)

  # Extract the top k frequent elements from the heap
  result = []
  while heap:
    result.append(heapq.heappop(heap)[1])

  # Return the result in any order
  return result

# Test cases
print(topKFrequent([1,1,1,2,2,3], 2))  # Output: [1, 2]
print(topKFrequent([1], 1))            # Output: [1]
```

###***26. Encode and Decode Strings (Medium)***

https://www.lintcode.com/problem/659/

```python
class Solution:
  def encode(self, strs):
    res = ""
    for s in strs:
      res += str(len(s)) + "#" + s
    return res

  def decode(self, str):
    res = []
    i = 0

    while i < len(str):
      j = i
      while str[j] != "#":
        j += 1
      length = int(str[i:j])
      res.append(str[j + 1: j + 1 + length])
      i = j + 1 + length
    return res

sol = Solution()
strs = ["Soe", "Wunna", "1996"]
x = sol.encode(strs)
print(x)
print(sol.decode(x))
```

###***27. Product of Array Except Self (Medium)***
https://leetcode.com/problems/product-of-array-except-self/

```python
from typing import List

# Prefix, Postfix problem

def product_except_self(nums: List[int]) -> List[int]:
    res = [1] * (len(nums))

    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]
    postfix = 1
    for i in range(len(nums) - 1, -1, -1):  # counting backward, 4 3 2 1 0
        res[i] *= postfix
        postfix *= nums[i]
    return res

print(product_except_self([1, 2, 3, 4]))
```

###***28. Valid Sudoku (Medium)***
https://leetcode.com/problems/valid-sudoku/

```python
from typing import List
import collections

class Solution:
  def isValidSudoku(self, board: List[List[str]]) -> bool:
    cols = collections.defaultdict(set)
    # if there is no key, key-value pair will be added itself when you add it
    rows = collections.defaultdict(set)
    squares = collections.defaultdict(set) # key = (r/3, c/3)

    for r in range(9): # 3x3 is already given
      for c in range(9):
        if board[r][c] == ".":
          continue
        if (board[r][c] in rows[r] or # check duplicate in rows
            board[r][c] in cols[c] or # check duplicate in cols
            board[r][c] in squares[(r // 3, c // 3)]): # check in 3x3 grid
          return False
        cols[c].add(board[r][c])
        rows[r].add(board[r][c])
        squares[(r // 3, c // 3)].add(board[r][c])
    return True
```

###***29. Longest Consecutive Sequence (Medium)***
https://leetcode.com/problems/longest-consecutive-sequence/

```python
from typing import List

class Solution:
  def longestConsecutive(self, nums: List[int]) -> int:
    numSet = set(nums)
    longest = 0

    for n in nums:
      # Check if it is the start of the sequence
      if (n - 1) not in numSet:
        length = 0
        while (n + length) in numSet:
          length += 1
        longest = max(length, longest)
    return longest
```

###***30. Sort Colors (Medium)***
https://leetcode.com/problems/sort-colors/

```python
from typing import List

class Solution:
  def sortColors(self, nums: List[int]) -> None:
    low = 0
    high = len(nums) - 1
    mid = 0

    while mid <= high:
      if nums[mid] == 0:
        nums[low], nums[mid] = nums[mid], nums[low]
        low += 1
        mid += 1
      elif nums[mid] == 1:
        mid += 1
      else:
        nums[mid], nums[high] = nums[high], nums[mid]
        high -= 1
    return nums

# sort() function runs in O(n*log n), timsort (merge sort + insertion sort)
```

###***31. Encode and Decode TinyURL (Medium)***
https://leetcode.com/problems/encode-and-decode-tinyurl/

```python
class Codec:
  def __init__(self):
    self.encodeMap = {}
    self.decodeMap = {}
    self.base = "http://tinyurl.com/"

  def encode(self, longUrl: str) -> str:
    """Encodes a URL to a shortened URL.
    """
    if longUrl not in self.encodeMap:
      shortUrl = self.base + str(len(self.encodeMap) + 1)
      self.encodeMap[longUrl] = shortUrl
      self.decodeMap[shortUrl] = longUrl
    return self.encodeMap[longUrl]

  def decode(self, shortUrl: str) -> str:
    """Decodes a shortened URL to its original URL.
    """
    return self.decodeMap[shortUrl]
```

###***32. Brick Wall (Medium)***
https://leetcode.com/problems/brick-wall/

```python
from typing import List

class Solution:
  def leastBricks(self, wall: List[List[int]]) -> int:
    countGap = {0:0} # { Position : Gap count }

    for r in wall:
      total = 0 # Current position
      for b in r[:-1]: # Exclude last brick
        total += b # position will be shifted
        countGap[total] = 1 + countGap.get(total, 0)

    return len(wall) - max(countGap.values()) # total number of rows - max gaps
```

```python
from typing import List

class Solution:
  def subarraySum(self, nums: List[int], k: int) -> int:
    res = 0
    curSum = 0
    prefixSums = { 0 : 1 }

    for n in nums:
      curSum += n
      diff = curSum - k

      res += prefixSums.get(diff, 0)
      prefixSums[curSum] = 1 + prefixSums.get(curSum, 0)

    return res
```

###***36. Minimum Number of Swaps to Make The String Balanced (Medium)***

```python
class Solution:
  def minSwaps(self, s: str) -> int:
    # Count extra number of closing brackets to calculate swap
    close, maxClose = 0, 0

    for c in s:
      if c == "[":
        close -= 1
      else:
        close += 1

      maxClose = max(maxClose, close) # Keeping track of total extra closing bracket

    return (maxClose + 1) // 2 # Or math.ceil(maxClose / 2)
    # Each swap gets rid of 2 extra closing brackets
```

###***37. Number of Pairs of Interchangeable Rectangles (Medium)***

```python
from typing import List

class Solution:
  def interchangeableRectangles(self, rectangles: List[List[int]]) -> int:
    count = {}

    for w, h in rectangles:
      count[w / h] = 1 + count.get(w / h, 0)

    res = 0
    for c in count.values():
      if c > 1:
        res += (c * (c - 1)) // 2
    return res
```

###***38. Maximum Product of The Length of Two Palindromic Subsequences (Medium)***
https://leetcode.com/problems/maximum-product-of-the-length-of-two-palindromic-subsequences/

```python
class Solution:
  def maxProduct(self, s: str) -> int:
    N, pali = len(s), {} # bitmask : length

    for mask in range(1, 1 << N): # 1 << N == 2 ** N
      subseq = ""
      for i in range(N):
        if mask & (1 << i):
          subseq += s[i]
      if subseq == subseq[::-1]:
        pali[mask] = len(subseq)

    res = 0
    for m1 in pali:
      for m2 in pali:
        if m1 & m2 == 0: # disjoint
          res = max(res, pali[m1] * pali[m2])

    return res
```

###***39. Grid Game (Medium)***

```python
from typing import List

class Solution:
  def gridGame(self, grid: List[List[int]]) -> int:
    N = len(grid[0]) # No columns
    preRow1, preRow2 = grid[0].copy(), grid[1].copy()

    for i in range(1, N): # start from 1, prefix for 0 is already calculated
      preRow1[i] += preRow1[i - 1]
      preRow2[i] += preRow2[i - 1]

    res = float("inf")
    for i in range(N):
      top = preRow1[-1] - preRow1[i]
      bottom = preRow2[i - 1] if i > 0 else 0
      secondRobot = max(top, bottom)
      res = min(res, secondRobot)
    return res
```

###***40. Find All Anagrams in a String (Medium)***
https://leetcode.com/problems/find-all-anagrams-in-a-string/

```python
from typing import List

class Solution:
  def findAnagrams(self, s: str, p: str) -> List[int]:

    startIndex = 0 # left pointer
    pMap, sMap = {}, {}
    res = []

    for char in p:
      pMap[char] = 1 + pMap.get(char, 0)

    for i in range(len(s)):
      sMap[s[i]] = 1 + sMap.get(s[i], 0)

      if i >= len(p) - 1:
        if sMap == pMap: # Anagram is found
          res.append(startIndex)

        # If current character is in sMap, remove it and re-update the map.
        if s[startIndex] in sMap:
          sMap[s[startIndex]] -= 1
          if sMap[s[startIndex]] == 0: # Not in current window
            del sMap[s[startIndex]]
        startIndex += 1 # Similar to l += 1

    return res
```

###***41. Find The Index of The First Occurrence in a String (Easy)***

```python
class Solution:
  def strStr(self, haystack: str, needle: str) -> int:
    if needle == "":
      return 0
    lps = [0] * len(needle)

    prevLPS, i = 0, 1
    while i < len(needle):
      if needle[i] == needle[prevLPS]:
        lps[i] = prevLPS + 1
        prevLPS += 1
        i += 1
      elif prevLPS == 0:
        lps[i] = 0
        i += 1
      else:
        prevLPS = lps[prevLPS - 1]

    i = 0  # ptr for haystack
    j = 0  # ptr for needle
    while i < len(haystack):
      if haystack[i] == needle[j]:
        i, j = i + 1, j + 1
      else:
        if j == 0:
          i += 1
        else:
          j = lps[j - 1]
      if j == len(needle):
        return i - len(needle)
    return -1

# Knuth–Morris–Pratt KMP (Hard)
```

```python
class Solution:
  def wiggleSort(self, nums):
    for i in range(1, len(nums)): # start from 1, no previous val from 0
      if i % 2 == 1 and nums[i] < nums[i - 1]: # odd index, previous must < cur, or swap
        nums[i], nums[i - 1] = nums[i - 1], nums[i] # swap values

      if i % 2 == 1 and nums[i] < nums[i - 1]: # even index, previous must > cur, or swap
        nums[i], nums[i - 1] = nums[i - 1], nums[i] # swap values
    return nums
```

###***45. Push Dominoes (Medium)***
https://leetcode.com/problems/push-dominoes/

```python
class Solution:
  def pushDominoes(self, dominoes: str) -> str:
    dom = list(dominoes) # To edit in place
    q = collections.deque()
    for i, d in enumerate(dom): # only standing dom will be added to q
      if d != '.':
        q.append((i, d))

    while q:
      i, d = q.popleft()

      if d == 'L' and i > 0 and dom[i - 1] == '.':
        q.append((i - 1, 'L'))
        dom[i - 1] = 'L'
      elif d == 'R':
        if i + 1 < len(dom) and dom[i + 1] == '.':
          if i + 2 < len(dom) and dom[i + 2] == 'L':
            q.popleft()
          else:
            q.append((i + 1, 'R'))
            dom[i + 1] = 'R'

    return ''.join(dom)
```

###***47. Insert Delete Get Random O(1) (Medium)***
https://leetcode.com/problems/insert-delete-getrandom-o1/

```python
from random import choice

class RandomizedSet:
  def __init__(self):
    self.numMap = {}
    self.numList = []

  def insert(self, val: int) -> bool:
    if val in self.numMap:
      return False
    self.numMap[val] = len(self.numList) # this will be the index after appending val to the list
    self.numList.append(val)
    return True

  def remove(self, val: int) -> bool:
    if val not in self.numMap:
      return False
    idx, last_element = self.numMap[val], self.numList[-1] # swap the element to be removed with last element
    self.numList[idx], self.numMap[last_element] = last_element, idx
    self.numList.pop() # O(1), we already moved it to last
    del self.numMap[val]
    return True

  def getRandom(self) -> int:
    return choice(self.numList) # random.choice()
```

###***49. Range Sum Query 2D Immutable (Medium)***
https://leetcode.com/problems/range-sum-query-2d-immutable/

```python
class NumMatrix:
  def __init__(self, matrix: List[List[int]]):
    ROWS, COLS = len(matrix), len(matrix[0])
    self.sumMat = [[0] * (COLS + 1) for r in range(ROWS + 1)]

    for r in range(ROWS):
      prefix = 0
      for c in range(COLS):
        prefix += matrix[r][c]
        above = self.sumMat[r][c + 1]
        self.sumMat[r + 1][c + 1] = prefix + above

  def sumRegion(self, r1: int, c1: int, r2: int, c2: int) -> int:
    r1, c1, r2, c2 = r1 + 1, c1 + 1, r2 + 1, c2 + 1

    bottomRight = self.sumMat[r2][c2]
    above = self.sumMat[r1 - 1][c2]
    left = self.sumMat[r2][c1 - 1]
    topLeft = self.sumMat[r1 - 1][c1 - 1]

    return bottomRight - above - left + topLeft
```

###***50. Non Decreasing Array (Medium)***
https://leetcode.com/problems/non-decreasing-array/

```python
from typing import List

class Solution:
  def checkPossibility(self, nums: List[int]) -> bool:
    changed = False

    for i in range(len(nums) - 1):
      if nums[i] <= nums[i + 1]:
        continue
      if changed:
        return False

      # If not in non decreasing order, will decrease the left value
      if i == 0 or nums[i + 1] >= nums[i - 1]: # [5, 3, 6] (i - 1, i, i + 1)
        nums[i] = nums[i + 1]
      else:
        nums[i + 1] = nums[i]
      changed = True
    return True
```

###***51. First Missing Positive (Hard)***
https://leetcode.com/problems/first-missing-positive/

```python
from typing import List

class Solution:
  def firstMissingPositive(self, nums: List[int]) -> int:
    for i in range(len(nums)):
      if nums[i] < 0:
        nums[i] = 0

    for i in range(len(nums)):
      val = abs(nums[i])
      if 1 <= val <= len(nums):
        if nums[val - 1] > 0:
          nums[val - 1] *= -1
        elif nums[val - 1] == 0:
          nums[val - 1] = -1 * (len(nums) + 1)

    for i in range(1, len(nums) + 1):
      if nums[i - 1] >= 0:
        return i
    return len(nums) + 1
```

###***57. Design Underground System (Medium)***
https://leetcode.com/problems/design-underground-system/

```python
class UndergroundSystem:
  def __init__(self):
    self.checkInMap = {}  # id -> (startStation, time)
    self.totalMap = {}  # (start, end) -> [totalTime, count]

  def checkIn(self, id: int, startStation: str, t: int) -> None:
    self.checkInMap[id] = (startStation, t)

  def checkOut(self, id: int, endStation: str, t: int) -> None:
    start, time = self.checkInMap[id]
    route = (start, endStation)
    if route not in self.totalMap:
      self.totalMap[route] = [0, 0]
    self.totalMap[route][0] += t - time
    self.totalMap[route][1] += 1

  def getAverageTime(self, startStation: str, endStation: str) -> float:
    total, count = self.totalMap[(startStation, endStation)]
    return total / count
```

###***58. Minimum Penalty for a Shop (Medium)***
https://leetcode.com/problems/minimum-penalty-for-a-shop/

```python
class Solution:
  def bestClosingTime(self, customers: str) -> int:
    n = len(customers)

    prefix_n = [0] * (n + 1) # N, extra value at the end
    postfix_y = [0] * (n + 1) # Y

    for i in range(1, n + 1): # Index zero is always zero/ handling out-of-bound
      prefix_n[i] = prefix_n[i - 1]
      if customers[i - 1] == "N":
        prefix_n[i] += 1

    for i in range(n - 1, -1, -1): # Skipping last val (n - 1)
      postfix_y[i] = postfix_y[i + 1] # opposite dir
      if customers[i] == "Y":
        postfix_y[i] += 1
    min_penalty, idx = float("inf"), 0

    for i in range(n + 1): # We are allowed to close at n
      penalty = prefix_n[i] + postfix_y[i]
      if penalty < min_penalty:
        min_penalty = penalty
        idx = i

    return idx
```

```python
class Solution:
  def bestClosingTime(self, customers: str) -> int:
    n = len(customers)
    penalty = 0
    for i in range(n):
      if customers[i] == 'Y':
        penalty += 1  # Penalty for open hour with customer
    min_penalty = penalty  # Initialize min_penalty to the penalty of closing at 0th hour
    optimal_closing_time = 0

    # Iterate through each possible closing time, updating penalty
    for closing_time in range(1, n + 1):
      if customers[closing_time - 1] == 'Y':
        penalty -= 1  # Adjust penalty for closing time with customer
      else:
        penalty += 1  # Adjust penalty for closing time with no customer

      # Update min_penalty and optimal_closing_time if needed
      if penalty < min_penalty:
        min_penalty = penalty
        optimal_closing_time = closing_time

    return optimal_closing_time
```

###***59. Text Justification (Hard)***
https://leetcode.com/problems/text-justification/

```python
from typing import List

class Solution:
  def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
    res = []
    line, length = [], 0
    i = 0

    while i < len(words):
      if length + len(line) + len(words[i]) > maxWidth:
        # Line complete
        extra_space = maxWidth - length
        spaces = extra_space // max(1, len(line) - 1)
        remainder = extra_space % max(1, len(line) - 1)

        for j in range(max(1, len(line) - 1)):
          line[j] += " " * spaces
          if remainder:
            line[j] += " "
            remainder -= 1
        res.append("".join(line))
        line, length = [], 0 # Reset line and length

      line.append(words[i])
      length += len(words[i])
      i += 1

    # Handling last line
    last_line = " ".join(line)
    trail_space = maxWidth - len(last_line)
    res.append(last_line + " " * trail_space)
    return res
```

###***60. Naming a Company (Hard)***
https://leetcode.com/problems/naming-a-company/

```python
from typing import List

class Solution:
  def distinctNames(self, ideas: List[str]) -> int:
    # map first char -> list of word suffixes
    wordMap = collections.defaultdict(set)
    for w in ideas:
      wordMap[w[0]].add(w[1:]) # Coffee, C -> offee

    res = 0
    for char1 in wordMap: # Every possible first char in wordMap
      for char2 in wordMap: # Loop again to create pair
        if char1 == char2:
          continue
        intersect = 0
        for w in wordMap[char1]: # w represents suffix here
          if w in wordMap[char2]:
            intersect += 1 # duplicates

        distinct1 = len(wordMap[char1]) - intersect
        distinct2 = len(wordMap[char2]) - intersect
        res += distinct1 * distinct2
    return res
```
