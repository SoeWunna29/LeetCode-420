###***1. Excel Sheet Column Title (Easy)***
https://leetcode.com/problems/excel-sheet-column-title/

```python
class Solution:
  def convertToTitle(self, columnNumber: int) -> str:
    # Time: O(logn) - Log base 26 of n
    res = ""
    while columnNumber > 0:
      remainder = (columnNumber - 1) % 26 # Subtracting 1 will give the correct number
      res += chr(ord('A') + remainder)
      columnNumber = (columnNumber - 1) // 26

    return res[::-1] # reverse output
```

###***2. Greatest Common Divisor of Strings (Easy)***
https://leetcode.com/problems/greatest-common-divisor-of-strings/

```python
class Solution:
  def gcdOfStrings(self, str1: str, str2: str) -> str:
    len1, len2 = len(str1), len(str2)

    def isDivisor(l):
      if len1 % l != 0 or len2 % l != 0: # non zero, not divisor
        return False
      X = str1[:l]
      return str1 == X * (len1 // l) and str2 == X * (len2 // l)

    for l in range(min(len1, len2), 0, -1): # Strat looping from end to index 0 of shorter len
      if isDivisor(l):
        return str1[:l]

    return ""
```

###***3. Count Odd Numbers in an Interval Range (Medium)***

```python
class Solution:
  def countOdds(self, low: int, high: int) -> int:
    length = high - low + 1
    count = length // 2
    if length % 2 and low % 2:
      count += 1
    return count
```

###***4. Matrix Diagonal Sum (Easy)***
https://leetcode.com/problems/matrix-diagonal-sum/

```python
from typing import List

class Solution:
  def diagonalSum(self, mat: List[List[int]]) -> int:
    res, n = 0, len(mat)

    for i in range(n):
      res += mat[i][i] # primary diagonal
      res += mat[i][len(mat) - 1 - i] # secondary diagonal

    return res - (mat[n // 2][n // 2] if n % 2 else 0)
```

###***6. Rotate Image (Medium)***
https://leetcode.com/problems/rotate-image/

```python
from typing import List

class Solution:
  def rotate(self, matrix: List[List[int]]) -> None:
    l, r = 0, len(matrix) - 1

    while l < r:
      for i in range(r - l):
        top, bottom = l, r # square matrix

        # save the topLeft
        topLeft = matrix[top][l + i]

        # move bottom left into top left
        matrix[top][l + i] = matrix[bottom - i][l]

        # move bottom right into bottom left
        matrix[bottom - i][l] = matrix[bottom][r - i]

        # move top right into bottom right
        matrix[bottom][r - i] = matrix[top + i][r]

        # move top left into top right
        matrix[top + i][r] = topLeft
      r -= 1
      l += 1
```

###***7. Spiral Matrix (Medium)***
https://leetcode.com/problems/spiral-matrix/

```python
from typing import List

class Solution:
  def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
    res = []
    left, right = 0, len(matrix[0]) # right is set out of bound for easier code
    top, bottom = 0, len(matrix) # same for bottom pointer

    while left < right and top < bottom:
      # get every i in the top row
      for i in range(left, right):
        res.append(matrix[top][i])
      top += 1

      # get every i in the right most col
      for i in range(top, bottom):
        res.append(matrix[i][right - 1])
      right -= 1

      if not (left < right and top < bottom):
        break

      # get every i in the bottom most row
      for i in range(right - 1, left - 1, -1): # from right to left/reverse loop
        res.append(matrix[bottom - 1][i])
      bottom -= 1

      # get every i in the left most col
      for i in range(bottom - 1, top - 1, -1): # bottom to top/reverse loop
        res.append(matrix[i][left])
      left += 1

    return res
```

###***8. Spiral Matrix II (Medium)***
https://leetcode.com/problems/spiral-matrix-ii/

```python
from typing import List

class Solution:
  def generateMatrix(self, n: int) -> List[List[int]]:
    mat = [[0] * n for _ in range(n)]

    left, right = 0, n - 1
    top, bottom = 0, n - 1
    val = 1

    while left <= right:
      # fill every val in top row
      for c in range(left, right + 1):
        mat[top][c] = val
        val += 1
      top += 1

      # fill every val in right col
      for r in range(top, bottom + 1):
        mat[r][right] = val
        val += 1
      right -= 1

      # fill every val in bottom row (reverse order)
      for c in range(right, left - 1, -1):
        mat[bottom][c] = val
        val += 1
      bottom -= 1

      # fill every val in the left col (reverse order)
      for r in range(bottom, top - 1, -1):
        mat[r][left] = val
        val += 1
      left += 1
    return mat
```

###***9. Set Matrix Zeroes (Medium)***
https://leetcode.com/problems/set-matrix-zeroes/

```python
from typing import List

class Solution:
  def setZeroes(self, matrix: List[List[int]]) -> None:
    # O(1)
    ROWS, COLS = len(matrix), len(matrix[0])
    rowZero = False

    # Determine which rows/cols need to be zero
    for r in range(ROWS):
      for c in range(COLS):
        if matrix[r][c] == 0:
          matrix[0][c] = 0 # in first row
          if r > 0:
            matrix[r][0] = 0 # starts after 1st row, in first col
          else:
            rowZero = True

      for r in range(1, ROWS):
        for c in range(1, COLS):
          if matrix[0][c] == 0 or matrix[r][0] == 0:
            matrix[r][c] = 0

      if matrix[0][0] == 0:
        for r in range(ROWS):
          matrix[r][0] = 0

      if rowZero:
        for c in range(COLS):
          matrix[0][c] = 0
```

###***10. Happy Number (Easy)***
https://leetcode.com/problems/happy-number/

```python
class Solution:
  def isHappy(self, n: int) -> bool:
    slow, fast = n, self.sumSquareDigits(n)

    while slow != fast: # Fast and Slow pointers solution
      fast = self.sumSquareDigits(fast) # 2 times
      fast = self.sumSquareDigits(fast) # logic -> fast = fast.next.next
      slow = self.sumSquareDigits(slow) # logic -> slow = slow.next

    return True if fast == 1 else False

  def sumSquareDigits(self, n):
    output = 0
    while n:
      output += (n % 10) ** 2 # 19 % 2 = 9
      n = n // 10 # 19 // 2 = 1
    return output
```

###***11. Plus One (Easy)***
https://leetcode.com/problems/plus-one/

```python
from typing import List

class Solution:
  def plusOne(self, digits: List[int]) -> List[int]:
    one = 1 # carry
    i = 0 # index
    digits = digits[::-1]

    while one:
      if i < len(digits):
        if digits[i] == 9: # 9 + 1 = 10, carry is set to 1, digit is set to 0
          digits[i] = 0
        else:
          digits[i] += 1 # just increase digit
          one = 0
      else:
        digits.append(one) # 999 + 1 becomes 1000
        one = 0
      i += 1
    return digits[::-1]
```

```python
class Solution:
  def isPalindrome(self, x: int) -> bool:
    if x < 0: return False

    div = 1
    while x >= 10 * div: # Isolate leftmost and rightmost digits of x
      div *= 10

    while x:
      right = x % 10 # 121 % 10 = 1, 12(1)
      left = x // div # 121 // 100 = 1, (1) 21

      if left != right: return False # not palindrome

      x = (x % div) // 10 # Remove the leftmost and rightmost digits from x.
      div = div / 100 # after removing eft and right), div // 100 to adjust for the reduction in the number of digits.
    return True
```

###***13. Ugly Number (Easy)***
https://leetcode.com/problems/ugly-number/

```python
class Solution:
  def isUgly(self, n: int) -> bool:
    if n <= 0:
      return False

    for p in [2, 3, 5]:
      while n % p == 0: #  removes all occurrences of p, prime factor from n
        n = n // p
    return n == 1
```

###***14. Shift 2D Grid (Easy)***
https://leetcode.com/problems/shift-2d-grid/

```python
from typing import List

class Solution:
  def shiftGrid(self, grid: List[List[int]], k: int) -> List[List[int]]:
    M, N = len(grid), len(grid[0])

    def posToVal(r, c): # Position to value
      return r * N + c

    def valToPos(v): # Value to position
      return [v // N, v % N]  # r, c

    res = [[0] * N for i in range(M)]
    for r in range(M):
      for c in range(N):
        newVal = (posToVal(r, c) + k) % (M * N)
        newR, newC = valToPos(newVal)
        res[newR][newC] = grid[r][c]

    return res
# Transform matrix into 1D array
```

###***15. Roman to Integer (Easy)***
https://leetcode.com/problems/roman-to-integer/

```python
# Roman to Integer

class Solution:
  def romanToInt(self, s: str) -> int:
    # largest t0 smallest: add them up
  # smaller before larget: subtract smaller
    roman = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    res = 0
    for i in range(len(s)):
      # In bound? and roman at i is < roman at i + 1?
      if i + 1 < len(s) and roman[s[i]] < roman[s[i + 1]]:
        res -= roman[s[i]]
      else:
        res += roman[s[i]]
    return res
```

###***16. Integer to Roman (Medium)***
https://leetcode.com/problems/integer-to-roman/

```python
# Integer to Roman

class Solution:
  def intToRoman(self, num: int) -> str:
    symList = [["I", 1], ["IV", 4], ["V", 5], ["IX", 9], ["X", 10], ["XL", 40],
               ["L", 50], ["XC", 90], ["C", 100], ["CD", 400], ["D", 500],
               ["CM", 900], ["M", 1000]]
    res = ""
    for sym, val in reversed(symList): # start dividing with largest val
      if num // val: # 40 // 1000 < 0, not possible for M to be in string, move to next roman
        count = num // val # count is how many times this roman letter will be in string
        res += sym * count
        num = num % val # update the num, take remainder
    return res
```

###***17. Pow(x, n) (Medium)***
https://leetcode.com/problems/powx-n/

```python
class Solution:
  def myPow(self, x: float, n: int) -> float:
    def helper(x, n):
      if x == 0: # 0 ** any is 0
        return 0
      if n == 0: # any ** 0 is 1
        return 1

      res = helper(x * x, n // 2)
      return x * res if n % 2 else res

    res = helper(x, abs(n))
    return res if n >= 0 else 1 / res # if exponent is negative -> 1 / res
```

###***18. Multiply Strings (Medium)***
https://leetcode.com/problems/multiply-strings/

```python
class Solution:
  def multiply(self, num1: str, num2: str) -> str:
    if "0" in [num1, num2]:
      return "0"

    res = [0] * (len(num1) + len(num2)) # 99 * 99 = 9801, can't exceed 4 digits
    num1, num2 = num1[::-1], num2[::-1] # reverse the string
    for i1 in range(len(num1)):
      for i2 in range(len(num2)):
        digit = int(num1[i1]) * int(num2[i2])
        res[i1 + i2] += digit
        res[i1 + i2 + 1] += res[i1 + i2] // 10
        res[i1 + i2] = res[i1 + i2] % 10

    res, beg = res[::-1], 0
    while beg < len(res) and res[beg] == 0: # removing the leading 0
      beg += 1 # increasing beginning pointer
    # array to string
    res = map(str, res[beg:])
    return "".join(res)
```

###***19. Detect Squares (Medium)***
https://leetcode.com/problems/detect-squares/

```python
from typing import List

class DetectSquares:
  def __init__(self):
    self.ptsCount = defaultdict(int)
    self.pts = []

  def add(self, point: List[int]) -> None:
    self.ptsCount[tuple(point)] += 1 # list can't be key
    self.pts.append(point)

  def count(self, point: List[int]) -> int:
    res = 0
    px, py = point
    for x, y in self.pts:
      # check diagonal distance
      if (abs(py - y) != abs(px - x)) or x == px or y == py:
        continue
      res += self.ptsCount[(x, py)] * self.ptsCount[(px, y)]
    return res
```

###***20. Robot Bounded In Circle (Medium)***
https://leetcode.com/problems/robot-bounded-in-circle/

```python
class Solution:
  def isRobotBounded(self, instructions: str) -> bool:
    # Directions: North, East, South, West
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    x, y = 0, 0  # Initial position
    dir_idx = 0  # Index for directions (0 -> North, 1 -> East, 2 -> South, 3 -> West)

    for instruction in instructions:
      if instruction == 'G':
        x += directions[dir_idx][0]
        y += directions[dir_idx][1]
      elif instruction == 'L':
        dir_idx = (dir_idx - 1) % 4  # Turn left: counter-clockwise
      elif instruction == 'R':
        dir_idx = (dir_idx + 1) % 4  # Turn right: clockwise

    # After one cycle of instructions:
    # Check if robot returns to origin or if it is not facing North
    return (x == 0 and y == 0) or (dir_idx != 0)
```

```python
class Solution:
  def isRobotBounded(self, instructions: str) -> bool:
    dirX, dirY = 0, 1 # 1 means facing north
    x, y = 0, 0 # starts at origin

    for d in instructions:
      if d == "G":
        x, y = x + dirX, y + dirY
      elif d == "L":
        dirX, dirY = -1 * dirY, dirX
      else: # Right direction
        dirX, dirY = dirY, -1 * dirX

    return (x, y) == (0, 0) or (dirX, dirY) != (0, 1)
```

###***21. Zigzag Conversion (Medium)***
https://leetcode.com/problems/zigzag-conversion/

```python
class Solution:
  def convert(self, s: str, numRows: int) -> str:
    if numRows == 1: return s # Base Case

    res = ""
    for r in range(numRows):
      increment = 2 * (numRows - 1)
      for i in range(r, len(s), increment):
        res += s[i]
        if (r > 0 and r < numRows - 1 and i + increment - 2 * r < len(s)):
          res += s[i + increment - 2 * r]

    return res
```

###***22. Find Missing Observations (Medium)***
https://leetcode.com/problems/find-missing-observations/

```python
from typing import List

class Solution:
  def missingRolls(self, rolls: List[int], mean: int, n: int) -> List[int]:
    m = len(rolls)
    nTotal = (mean * (n + m)) - sum(rolls) # entire total - m total

    if nTotal < n or nTotal > n * 6:
      return []

    res = []
    while nTotal:
      dice = min(nTotal - n + 1, 6) # Greedy
      res.append(dice)
      nTotal -= dice
      n -= 1
    return res
```
