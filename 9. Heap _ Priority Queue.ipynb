###***1. Kth Largest Element In a Stream (Easy)***
https://leetcode.com/problems/kth-largest-element-in-a-stream/

```python
from typing import List
import heapq

class KthLargest:
  def __init__(self, k: int, nums: List[int]):
    self.minHeap, self.k = nums, k
    heapq.heapify(self.minHeap)
    while len(self.minHeap) > k:
      heapq.heappop(self.minHeap)

  def add(self, val: int) -> int:
    heapq.heappush(self.minHeap, val)
    if len(self.minHeap) > self.k:
      heapq.heappop(self.minHeap)
    return self.minHeap[0]
```

###***2. Last Stone Weight (Easy)***
https://leetcode.com/problems/last-stone-weight/

```python
from typing import List

class Solution:
  def lastStoneWeight(self, stones: List[int]) -> int:
    stones = [-s for s in stones] # max heap
    heapq.heapify(stones)

    while len(stones) > 1:
      first = -1 * heapq.heappop(stones)
      second = -1 * heapq.heappop(stones)
      if second < first: # first stone is always bigger/ max heap
        heapq.heappush(stones, -1 * (first - second))

    stones.append(0) # edge case if stones is empty
    return abs(stones[0])
```

###***3. K Closest Points to Origin (Medium)***
https://leetcode.com/problems/k-closest-points-to-origin/

```python
from typing import List

class Solution:
  def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
    minHeap = []
    for x, y in points:
      dist = (x ** 2) + (y ** 2)
      minHeap.append([dist, x, y])

    heapq.heapify(minHeap)
    res = []
    while k > 0:
      dist, x, y = heapq.heappop(minHeap)
      res.append([x, y])
      k -= 1

    return res
```

###***5. Task Scheduler (Medium)***
https://leetcode.com/problems/task-scheduler/

```python
from typing import List

class Solution:
  def leastInterval(self, tasks: List[str], n: int) -> int:
    # each task 1 unit time
    # minimize idle time

    count = Counter(tasks)
    maxHeap = [-cnt for cnt in count.values()]
    heapq.heapify(maxHeap)

    time = 0
    q = deque() # pairs of [-cnt, idleTime]
    while maxHeap or q:
      time += 1 # time will be increased every iteration
      if maxHeap:
        cnt = 1 + heapq.heappop(maxHeap) # to increase negative value
        if cnt:
          q.append([cnt, time + n])
      if q and q[0][1] == time: # can perform that task again at q[0][1] == time
        heapq.heappush(maxHeap, q.popleft()[0])
    return time
```

###***6. Design Twitter (Medium)***
https://leetcode.com/problems/design-twitter/

```python
from typing import List

class Twitter:
  def __init__(self):
    self.count = 0
    self.tweetMap = defaultdict(list) # userId -> list of [count, tweetId]
    self.followMap = defaultdict(set) # userId -> set of followeeId

  def postTweet(self, userId: int, tweetId: int) -> None:
    self.tweetMap[userId].append([self.count, tweetId])
    self.count -= 1

  def getNewsFeed(self, userId: int) -> List[int]:
    res = []
    minHeap = []

    self.followMap[userId].add(userId)
    for followeeId in self.followMap[userId]:
      if followeeId in self.tweetMap:
        index = len(self.tweetMap[followeeId]) - 1
        count, tweetId = self.tweetMap[followeeId][index]
        minHeap.append([count, tweetId, followeeId, index - 1])
    heapq.heapify(minHeap)

    while minHeap and len(res) < 10:
      count, tweetId, followeeId, index = heapq.heappop(minHeap)
      res.append(tweetId)
      if index >= 0:
        count, tweetId = self.tweetMap[followeeId][index]
        heapq.heappush(minHeap, [count, tweetId, followeeId, index - 1])
    return res

  def follow(self, followerId: int, followeeId: int) -> None:
    self.followMap[followerId].add(followeeId)

  def unfollow(self, followerId: int, followeeId: int) -> None:
    if followeeId in self.followMap[followerId]:
      self.followMap[followerId].remove(followeeId)
```

###***7. Minimize Deviation in Array (Hard)***
https://leetcode.com/problems/minimize-deviation-in-array/


```python
from typing import List

class Solution:
  def minimumDeviation(self, nums: List[int]) -> int:
    minHeap, heapMax = [], 0
    for n in nums:
      tmp = n
      while n % 2 == 0:
        n = n // 2
      minHeap.append((n, max(tmp, 2 * n)))
      heapMax = max(heapMax, n)

    res = float("inf")
    heapq.heapify(minHeap)
    while len(minHeap) == len(nums):
      n, nMax = heapq.heappop(minHeap)
      res = min(res, heapMax - n)

      if n < nMax:
        heapq.heappush(minHeap, (n * 2, nMax))
        heapMax = max(heapMax, n * 2)
    return res
```

```python
from typing import List

class Solution:
  def minimumDeviation(self, nums: List[int]) -> int:
    minHeap, heapMax = [], 0

    # Preprocess nums, dividing even numbers by 2 until they become odd
    for n in nums:
      tmp = n
      while n % 2 == 0:
        n = n // 2
      minHeap.append((n, max(tmp, 2 * n)))
      heapMax = max(heapMax, n)

    res = float("inf")
    heapq.heapify(minHeap)

    # Continue processing until minHeap contains all elements from nums
    while len(minHeap) == len(nums):
      n, nMax = heapq.heappop(minHeap)
      res = min(res, heapMax - n) # Update the result with the minimum deviation

      if n < nMax: # If the processed number is less than its maximum value
        heapq.heappush(minHeap, (n * 2, nMax)) # Insert the doubled value and its maximum into minHeap
        heapMax = max(heapMax, n * 2)

    return res
```

###***8. Maximum Subsequence Score (Medium)***
https://leetcode.com/problems/maximum-subsequence-score/

```python
from typing import List

class Solution:
  def maxScore(self, nums1: List[int], nums2: List[int], k: int) -> int:
    pairs = [(n1, n2) for n1, n2 in zip(nums1, nums2)]
    pairs = sorted(pairs, key=lambda p:p[1], reverse=True)

    minHeap = []
    n1Sum = 0
    res = 0

    for n1, n2 in pairs:
      n1Sum += n1
      heapq.heappush(minHeap, n1)

      if len(minHeap) > k:
        n1Pop = heapq.heappop(minHeap)
        n1Sum -= n1Pop
      if len(minHeap) == k:
        res = max(res, n1Sum * n2)
    return res
```

###***9. Single Threaded Cpu (Medium)***
https://leetcode.com/problems/single-threaded-cpu/

```python
from typing import List
import heapq

class Solution:
  def getOrder(self, tasks: List[List[int]]) -> List[int]:
    # Add index to each task for tracking original order
    for i, t in enumerate(tasks):
      t.append(i)
    # Sort tasks by their start time
    tasks.sort(key=lambda t:t[0])

    res, minHeap = [], []
    i, time = 0, tasks[0][0]

    # Process tasks
    while minHeap or i < len(tasks):
      # Push tasks into heap that can be processed at current time
      while i < len(tasks) and time >= tasks[i][0]:
        heapq.heappush(minHeap, [tasks[i][1], tasks[i][2]])  # Push task [processing time, index] into heap
        i += 1

      # If no tasks in heap, move time to next task start time
      if not minHeap:
        time = tasks[i][0]
      else:
        # Pop task from heap and process it
        procTime, index = heapq.heappop(minHeap)
        time += procTime  # Update current time
        res.append(index)  # Append index of processed task to result list

    return res
```

###***11. Process Tasks Using Servers (Medium)***
https://leetcode.com/problems/process-tasks-using-servers/

```python
from typing import List

class Solution:
  def assignTasks(self, servers: List[int], tasks: List[int]) -> List[int]:
    res = [0] * len(tasks)
    available = [(servers[i], i) for i in range(len(servers))]
    heapq.heapify(available)
    unavail = []

    t = 0
    for i in range(len(tasks)):
      t = max(t, i)

      if len(available) == 0:
        t = unavail[0][0]
      while unavail and t >= unavail[0][0]:
        timefree, weight, index = heapq.heappop(unavail)
        heapq.heappush(available, (weight, index))

      weight, index = heapq.heappop(available)
      res[i] = index
      heapq.heappush(unavail, (t + tasks[i], weight, index))
    return res
```

```python
from typing import List
import heapq

class Solution:
  def assignTasks(self, servers: List[int], tasks: List[int]) -> List[int]:
    # Initialize a list to store the result of assigned tasks
    res = [0] * len(tasks)

    # Create a list of tuples containing server weight and index
    available = [(servers[i], i) for i in range(len(servers))]

    # Convert the list into a min-heap based on server weights
    heapq.heapify(available)

    # List to store unavailable servers along with their time of availability,
    # task weight, and index
    unavail = []

    # Initialize time
    t = 0

    # Iterate over each task
    for i in range(len(tasks)):
      # Ensure time is at least equal to the current task index
      t = max(t, i)

      # If no servers are available, update time to the earliest availability
      if len(available) == 0:
        t = unavail[0][0]

      # Assign tasks to available servers and update time if servers are unavailable
      while unavail and t >= unavail[0][0]:
        timefree, weight, index = heapq.heappop(unavail)
        heapq.heappush(available, (weight, index))

      # Assign the task to the server with the lowest weight
      weight, index = heapq.heappop(available)
      res[i] = index

      # Add the server to the list of unavailable servers along with its availability time
      heapq.heappush(unavail, (t + tasks[i], weight, index))

    return res
```

###***12. Find The Kth Largest Integer In The Array (Medium)***
https://leetcode.com/problems/find-the-kth-largest-integer-in-the-array/

```python
from typing import List

class Solution:
  def kthLargestNumber(self, nums: List[str], k: int) -> str:
    maxHeap = [-int(n) for n in nums]
    heapq.heapify(maxHeap)

    while k > 1:
      heapq.heappop(maxHeap)
      k -= 1
    return str(-maxHeap[0])
```

###***13. Reorganize String (Medium)***
https://leetcode.com/problems/reorganize-string/

```python
class Solution:
  def reorganizeString(self, s: str) -> str:
    count = Counter(s)  # Hashmap, count each char
    maxHeap = [[-cnt, char] for char, cnt in count.items()]
    heapq.heapify(maxHeap)  # O(n)

    prev = None
    res = ""
    while maxHeap or prev:
      if prev and not maxHeap: # Couldn't build string
        return ""

      # most frequent, except prev
      cnt, char = heapq.heappop(maxHeap)
      res += char
      cnt += 1 # Decrementing negative count (-cnt)

      if prev: # If prev, add it back to maxHeap
        heapq.heappush(maxHeap, prev)
        prev = None # Set it to none

      if cnt != 0: # Setting prev for next iteration
        prev = [cnt, char]
    return res
```

###***14. Longest Happy String (Medium)***
https://leetcode.com/problems/longest-happy-string/

```python
class Solution:
  def longestDiverseString(self, a: int, b: int, c: int) -> str:
    res, maxHeap = "", [] # Using MaxHeap to start adding most frequent char
    for count, char in [(-a, "a"), (-b, "b"), (-c, "c")]:
      if count != 0: # To avoid edge case
        heapq.heappush(maxHeap, (count, char))

    while maxHeap:
      count, char = heapq.heappop(maxHeap)
      if len(res) > 1 and res[-1] == res[-2] == char: # "aa(char)" res[-1] == res[-2]
        if not maxHeap:
          break
        count2, char2 = heapq.heappop(maxHeap) # Get new char and count
        res += char2
        count2 += 1 # Decrementing negative val count
        if count2: # If count2 is not empty yet
          heapq.heappush(maxHeap, (count2, char2))

      else:
        res += char
        count += 1 # Decrementing negative val count

      if count: # If count is not empty yet
        heapq.heappush(maxHeap, (count,char))
    return res
```

###***15. Car Pooling (Medium)***
https://leetcode.com/problems/car-pooling/

```python
from typing import List

class Solution:
  def carPooling(self, trips: List[List[int]], capacity: int) -> bool:
    # O(nlogn)
    trips.sort(key=lambda t:t[1]) # Sort based on starting val [no, start, end]

    minHeap = [] # pair of [end, numPassenger]
    curPass = 0
    for t in trips:
      numPass, start, end = t
      while minHeap and minHeap[0][0] <= start: # Completed trips?, end <= start?
        curPass -= minHeap[0][1] # Dropped off passengers
        heapq.heappop(minHeap)

      curPass += numPass
      if curPass > capacity:
        return False
      heapq.heappush(minHeap, [end, numPass])

    return True
```

###***16. Find Median From Data Stream (Hard)***
https://leetcode.com/problems/find-median-from-data-stream/

```python
class MedianFinder:
  def __init__(self):
    # Two heaps -> Large, Small -> Min-Heap, Max-Heap
    # Heaps should be of equal size
    self.small, self.large = [], []

  def addNum(self, num: int) -> None:
    heapq.heappush(self.small, -1 * num) # making max num become min

    # Make sure every num in small is <= every num in large
    if (self.small and self.large and
        (-1 * self.small[0]) > self.large[0]):
      val = -1 * heapq.heappop(self.small) # Converting back to max num
      heapq.heappush(self.large, val)

    # Uneven size?
    if len(self.small) > len(self.large) + 1:
      val = - 1 * heapq.heappop(self.small)
      heapq.heappush(self.large, val)
    if len(self.large) > len(self.small) + 1:
      val = heapq.heappop(self.large)
      heapq.heappush(self.small, -1 * val)

  def findMedian(self) -> float:
    if len(self.small) > len(self.large):
      return -1 * self.small[0]
    if len(self.large) > len(self.small):
      return self.large[0]

    return (-1 * self.small[0] + self.large[0]) / 2
```

###***17. Maximum Performance of a Team (Hard)***
https://leetcode.com/problems/maximum-performance-of-a-team/

```python
from typing import List

class Solution:
  def maxPerformance(self, n: int, speed: List[int], efficiency: List[int], k: int) -> int:
    eng = []
    for eff, spd in zip(efficiency, speed):
      eng.append([eff, spd])
    eng.sort(reverse=True) # Descending order -> efficiency

    res, speed = 0, 0
    minHeap = []
    for eff, spd in eng:
      if len(minHeap) == k:
        speed -= heapq.heappop(minHeap) # Deduct min speed from total speed

      speed += spd # Total speed
      heapq.heappush(minHeap, spd) # Add to minHeap to pop lowest speed later

      # Since we sorted in reverse order, no need to look back greater efficiency
      res = max(res, eff * speed) # Current eff will always be minimum so far

    return res % (10 ** 9 + 7) # According to description
```

```python
import heapq

def maxPerformance(n, speed, efficiency, k):
  MOD = 10 ** 9 + 7
  engineers = sorted(zip(efficiency, speed), reverse=True) # Sort by efficiency descending
  min_efficiency_heap = [] # Min heap for efficiency
  speed_sum = 0
  max_performance = 0

  for eff, spd in engineers:
    heapq.heappush(min_efficiency_heap, spd)
    speed_sum += spd

    if len(min_efficiency_heap) > k:
      speed_sum -= heapq.heappop(min_efficiency_heap)

    max_performance = max(max_performance, speed_sum * eff)

  return max_performance % MOD
```

###***18. IPO (Hard)***
https://leetcode.com/problems/ipo/

```python
from typing import List

class Solution:
  def findMaximizedCapital(self, k: int, w: int, profits: List[int], capital: List[int]) -> int:
    # Two Heaps
    maxProfit = [] # only projects we can afford
    minCapital = [(c, p) for c, p in zip(capital, profits)]
    heapq.heapify(minCapital) # Sort by min capital

    for i in range(k): # At most k projects
      while minCapital and minCapital[0][0] <= w: # Profits with available capital
        c, p = heapq.heappop(minCapital)
        heapq.heappush(maxProfit, -1 * p) # Adding profit to max heap
      if not maxProfit: # If profit heap is empty
        break
      w += -1 * heapq.heappop(maxProfit) # Adding max  profit to initial capital
    return w
```
