###***1. Valid Palindrome (Easy)***
https://leetcode.com/problems/valid-palindrome/

```python
class Solution:
  def is_palindrome(self, s: str) -> bool:
    # Convert to lowercase and remove non-alphanumeric characters
    cleaned_s = ''.join(char.lower() for char in s if char.isalnum())

    # Check if the cleaned string is equal to its reverse
    return cleaned_s == cleaned_s[::-1]

sol = Solution()
s = "A man, a plan, a canal: Panama"
print(sol.is_palindrome(s))
```

###***2. Valid Palindrome II (Easy)***
https://leetcode.com/problems/valid-palindrome-ii/

```python
from typing import List

class Solution:
  def validPalindrome(self, s: str) -> bool:
    l, r = 0, len(s) - 1

    while l < r:
      if s[l] == s[r]:
        l += 1
        r -= 1
      else: # Recursive call, (l + 1, r) or (l, r - 1), remove from left or right
        return self.validPalindromeUtil(s, l + 1, r) or self.validPalindromeUtil(s, l, r - 1)
    return True

  def validPalindromeUtil(self, s, l, r):
    while l < r:
      if s[l] == s[r]:
        l += 1
        r -= 1
      else:
        return False

    return True
```

###***3. Minimum Difference Between Highest And Lowest of K Scores (Easy)***
https://leetcode.com/problems/minimum-difference-between-highest-and-lowest-of-k-scores/

```python
from typing import List

class Solution:
  def minimumDifference(self, nums: List[int], k: int) -> int:
    nums.sort()
    l, r = 0, k - 1 # Sliding Window
    res = float("inf")

    while r < len(nums):
      res = min(res, nums[r] - nums[l])
      l, r = l + 1, r + 1
    return res
```

###***4. Merge Strings Alternately (Easy)***
https://leetcode.com/problems/merge-strings-alternately/

```python
class Solution:
  def mergeAlternately(self, word1: str, word2: str) -> str:
    i = j = 0
    res = []

    while i < len(word1) and j < len(word2):
      res.append(word1[i])
      res.append(word2[j])
      i += 1
      j += 1
    res.append(word1[i:])
    res.append(word2[j:])
    return ''.join(res)
```

###***5. Reverse String (Easy)***
https://leetcode.com/problems/reverse-string/

```python
from typing import List

class Solution:
  def reverseString(self, s: List[str]) -> None:
    l, r = 0, len(s) - 1
    while l < r:
      s[l], s[r] = s[r], s[l]
      l += 1
      r -= 1
```

###***6. Merge Sorted Array (Easy)***
https://leetcode.com/problems/merge-sorted-array/

```python
from typing import List

class Solution:
  def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
    # Last index of nums1
    last = m + n - 1 # [1, 2, 3, 0, 0, (0)]

    # Merge in reverse order
    while m > 0 and n > 0:
      if nums1[m - 1] > nums2[n - 1]: # Last values
        nums1[last] = nums1[m - 1]
        m -= 1
      else:
        nums1[last] = nums2[n - 1]
        n -= 1
      last -= 1 # last value is filled and move to prev index

    # fill num1 with leftover nums2 elements
    while n > 0:
      nums1[last] = nums2[n - 1]
      n, last = n - 1, last - 1
```

###***7. Move Zeroes (Easy)***
https://leetcode.com/problems/move-zeroes/

```python
from typing import List

class Solution:
  def moveZeroes(self, nums: List[int]) -> None:
    l = 0
    for r in range(len(nums)):
      if nums[r]: # Move positive nums to the start, instead of zeros to end
        nums[l], nums[r] = nums[r], nums[l]
        l += 1 # Updating left pointer
    return nums
```

###***8. Remove Duplicates From Sorted Array (Easy)***
https://leetcode.com/problems/remove-duplicates-from-sorted-array/

```python
from typing import List

class Solution:
  def removeDuplicates(self, nums: List[int]) -> int:
    l = 1

    for r in range(1, len(nums)): # Start from 1 so that previous val can be checked
      if nums[r] != nums[r - 1]:
        nums[l] = nums[r] # Replacing in place
        l += 1
    return l
```

###***9. Remove Duplicates From Sorted Array II (Medium)***
https://leetcode.com/problems/remove-duplicates-from-sorted-array-ii/

```python
from typing import List

class Solution:
  def removeDuplicates(self, nums: List[int]) -> int:
    l, r = 0, 0

    while r < len(nums):
      count = 1
      while r + 1 < len(nums) and nums[r] == nums[r + 1]:
        r += 1
        count += 1

      for i in range(min(2, count)):
        nums[l] = nums[r]
        l += 1
      r += 1
    return l
```

###***10. Two Sum II Input Array Is Sorted (Medium)***
https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/

```python
def two_sum2(numbers, target):
    l, r = 0, len(numbers) - 1

    while l < r:
        curSum = numbers[l] + numbers[r]

        if curSum > target:
            r -= 1
        elif curSum < target:
            l += 1
        else:
            return [l + 1, r + 1]
    return

numbers = [2, 7, 11, 15]
target = 9
print(two_sum2(numbers, target))
```

###***11. 3Sum (Medium)***
https://leetcode.com/problems/3sum/

```python
def three_sum(nums):
    res = []
    nums.sort()

    for i, a in enumerate(nums):
        if i > 0 and a == nums[i - 1]:
            continue

        l = i + 1
        r = len(nums) - 1
        while l < r:
            threeSum = a + nums[l] + nums[r]
            if threeSum > 0:
                r -= 1
            elif threeSum < 0:
                l += 1
            else:
                res.append([a, nums[l], nums[r]])
                l += 1
                while nums[l] == nums[l - 1] and l < r:
                    l += 1

    return res

nums = [-1, 0, 1, 2, -1, -4]
print(three_sum(nums))
```

###***12. 4Sum (Medium)***
https://leetcode.com/problems/4sum/

```python
from typing import List

class Solution:
  def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
    nums.sort()
    res, quad = [], []

    def kSum(k, start, target):
      if k != 2:
        for i in range(start, len(nums) - k + 1):
          if i > start and nums[i] == nums[i - 1]:
            continue
          quad.append(nums[i])
          kSum(k - 1, i + 1, target - nums[i])
          quad.pop()
        return
      # base case, two sum II
      l, r = start, len(nums) - 1
      while l < r:
        if nums[l] + nums[r] < target:
          l += 1
        elif nums[l] + nums[r] > target:
          r -= 1
        else:
          res.append(quad + [nums[l], nums[r]])
          l += 1
          while l < r and nums[l] == nums[l - 1]:
            l += 1

    kSum(4, 0, target)
    return res
```

###***13. Container With Most Water (Medium)***
https://leetcode.com/problems/container-with-most-water/

```python
height = [1, 8, 6, 2, 5, 4, 8, 3 ,7]
left = 0
right = len(height) - 1
area = 0

while left < right:
  area = max(area, (right - left) * min(height[left], height[right]))
  if height[left] < height[right]:
    left += 1
  else:
    right -= 1

print(area)
```

###***14. Number of Subsequences That Satisfy The Given Sum Condition (Medium)***
https://leetcode.com/problems/number-of-subsequences-that-satisfy-the-given-sum-condition/

```python
from typing import List

class Solution:
  def numSubseq(self, nums: List[int], target: int) -> int:
    nums.sort()
    res = 0
    mod = (10 ** 9 + 7)

    r = len(nums) - 1
    for i, left in enumerate(nums):
      while (left + nums[r]) > target and i <= r:
        r -= 1
      if i <= r:
        res += (2 **(r - i))
        res %= mod

    return res
```

###***15. Rotate Array (Medium)***
https://leetcode.com/problems/rotate-array/

```python
from typing import List

class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        if len(nums) == 0:
            return
        k = k % len(nums) # Don't rotate more times than necessary, if k > len(nums)
        pivot = (len(nums) - 1) - k # Pivot point to rotate
        nums[:] = nums[pivot + 1:] + nums[:pivot + 1] # Editing in place
```

###***16. Array With Elements Not Equal to Average of Neighbors (Medium)***
https://leetcode.com/problems/array-with-elements-not-equal-to-average-of-neighbors/

```python
from typing import List

class Solution:
  def rearrangeArray(self, nums: List[int]) -> List[int]:
    nums.sort()
    res = []

    l, r = 0, len(nums) - 1
    while len(res) != len(nums): # Will add alternatively to satisfy the given condition
      res.append(nums[l]) # Start from first
      l += 1

      if l <= r: # To avoid adding duplicate
        res.append(nums[r]) # From end
        r -= 1
    return res
```

###***17. Boats to Save People (Medium)***
https://leetcode.com/problems/boats-to-save-people/

```python
class Solution:
  def numRescueBoats(self, people: list[int], limit: int) -> int:
    people.sort()

    res = 0 # boats
    left, right = 0, len(people) - 1
    while left <= right:
      if people[left] + people[right] <= limit:
        left += 1
      right -= 1
      res += 1
    return res
```

###***18. Trapping Rain Water (Hard)***
https://leetcode.com/problems/trapping-rain-water/

```python
from typing import List

class Solution:
  def trap(self, height: List[int]) -> int:
    if not height: return 0

    l, r = 0, len(height) - 1
    leftMax, rightMax = height[l], height[r]
    res = 0

    while l < r:
      if leftMax < rightMax:
        l += 1
        leftMax = max(leftMax, height[l])
        res += leftMax - height[l]
      else:
        r -= 1
        rightMax = max(rightMax, height[r])
        res += rightMax - height[r]

    return res
```
