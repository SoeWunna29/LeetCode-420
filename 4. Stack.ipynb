###***1. Valid Parentheses (Easy)***
https://leetcode.com/problems/valid-parentheses/

```python
class Solution:
  def is_valid(self, s: str) -> bool:
    stack = []
    closeToOpen = {")": "(",
                   "]": "[",
                   "}": "{"}

    for c in s:
      if c in closeToOpen: # if c is the close bracket
        if stack and stack[-1] == closeToOpen[c]:
          stack.pop()
        else: # can't start with close bracket
          return False
      else: # if c is open bracket
        stack.append(c)

    return True if not stack else False

sol = Solution()
s = "()[]{}"
print(sol.is_valid(s))
```

###***2. Baseball Game (Easy)***
https://leetcode.com/problems/baseball-game/

```python
from typing import List

class Solution:
  def calPoints(self, operations: List[str]) -> int:
    stack = []

    for op in operations:
      if op == "+":
        stack.append(stack[-1] + stack[-2])
      elif op == "D":
        stack.append(2 * stack[-1])
      elif op == "C":
        stack.pop()
      else:
        stack.append(int(op))

    return sum(stack)
```

###***3. Implement Stack Using Queues (Easy)***
https://leetcode.com/problems/implement-stack-using-queues/

```python
from collections import deque

class MyStack:
  def __init__(self):
    self.q = deque()

  def push(self, x: int) -> None:
    self.q.append(x)

  def pop(self) -> int:
    for i in range(len(self.q) - 1): # [1, 2, 3] -> [3, 1, 2]
      self.push(self.q.popleft())
    return self.q.popleft() # [(3), 1, 2]

  def top(self) -> int: # One line -> return self.q[-1]
    for i in range(len(self.q) - 1): # [1, 2, 3] -> [3, 1, 2]
      self.push(self.q.popleft())
    res = self.q[0] # [(3), 1, 2]
    self.push(self.q.popleft()) # [1, 2, 3], resetting to original
    return res

  def empty(self) -> bool:
    return len(self.q) == 0
```

###***4. Min Stack (Medium)***
https://leetcode.com/problems/min-stack/

```python
class MinStack:
  def __init__(self):
    self.stack = []
    self.minStack = []

  def push(self, val: int) -> None:
    self.stack.append(val)
    val = min(val, self.minStack[-1] if self.minStack else val)
    self.minStack.append(val)

  def pop(self) -> None:
    self.stack.pop()
    self.minStack.pop()

  def top(self) -> int:
    return self.stack[-1]

  def getMin(self) -> int:
    return self.minStack[-1]
```

###***5. Evaluate Reverse Polish Notation (Medium)***
https://leetcode.com/problems/evaluate-reverse-polish-notation/

```python
from typing import List

class Solution:
  def evalRPN(self, tokens: List[str]) -> int:
    stack = []
    for c in tokens:
      if c == "+":
        stack.append(stack.pop() + stack.pop())
      elif c == "-":
        a, b = stack.pop(), stack.pop()
        stack.append(b - a)
      elif c == "*":
        stack.append(stack.pop() * stack.pop())
      elif c == "/":
        a, b = stack.pop(), stack.pop()
        stack.append(int(b / a))
      else:
        stack.append(int(c))
    return stack[0]
```

###***6. Removing Stars From a String (Medium)***
https://leetcode.com/problems/removing-stars-from-a-string/

```python
class Solution:
  def removeStars(self, s):
    res = []
    for c in s:
      if res and c == '*':
        res.pop()
      else:
        res.append(c)
    return ''.join(res)
```

###***7. Validate Stack Sequences (Medium)***
https://leetcode.com/problems/validate-stack-sequences/

```python
class Solution:
  def validateStackSequences(self, pushed, popped):
    i = 0 # Indexing popped list
    stack = []
    for n in pushed:
      stack.append(n)
      while i < len(popped) and stack and popped[i] == stack[-1]:
        stack.pop()
        i += 1 # Increase after every iteration

    return not stack # True if stack is empty

# Pushed must come first, popped[i] == stack[-1] (top of stack)
```

###***8. Generate Parentheses (Medium)***
https://leetcode.com/problems/generate-parentheses/

```python
class Solution:
  def generateParenthesis(self, n: int) -> List[str]:
    # only add open parenthesis if open < n
    # only add a closing parenthesis if closed < open
    # valid if open == closed == n

    stack = []
    res = []

    def backtrack(openN, closedN):
      if openN == closedN == n:
        res.append("".join(stack))
        return

      if openN < n:
        stack.append("(")
        backtrack(openN + 1, closedN)
        stack.pop()

      if closedN < openN:
        stack.append(")")
        backtrack(openN, closedN + 1)
        stack.pop()

    backtrack(0, 0)
    return res
```

###***9. Asteroid Collision (Medium)***
https://leetcode.com/problems/asteroid-collision/

```python
from typing import List

class Solution:
  def asteroidCollision(self, asteroids: List[int]) -> List[int]:
    stack = []

    for a in asteroids:
      while stack and a < 0 and stack[-1] > 0: # Stack is not empty
        diff = a + stack[-1]
        if diff < 0: # Size of a is bigger
          stack.pop()
        elif diff > 0: # a is destroyed
          a = 0 # Zero won't be added to stack
        else: # Equal size, both will be destroyed
          a = 0
          stack.pop()

      if a: # Only if a is not zero
        stack.append(a)

    return stack
```

###***10. Daily Temperatures (Medium)***
https://leetcode.com/problems/daily-temperatures/

```python
from typing import List

class Solution:
  def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
    res = [0] * len(temperatures)
    stack = [] # pair: [temp, index]

    for i, t in enumerate(temperatures): # index, temp
      while stack and t > stack[-1][0]:
        stackT, stackInd = stack.pop()
        res[stackInd] = (i - stackInd) # will replace in corresponding index
      stack.append([t, i])
    return res

# Monotonic decreasing stack
```

###***11.Online Stock Span (Medium)***
https://leetcode.com/problems/online-stock-span/

```python
class StockSpanner:
  def __init__(self):
    self.stack = []  # pair: (price, span)

  def next(self, price: int) -> int:
    span = 1 # Today price itself counts as 1
    while self.stack and self.stack[-1][0] <= price:
      span += self.stack[-1][1]
      self.stack.pop()

    self.stack.append((price, span))
    return span
```

###***12. Car Fleet (Medium)***
https://leetcode.com/problems/car-fleet/

```python
from typing import List

class Solution:
  def carFleet(self, target: int, position: List[int], speed: List[int]) -> int:
    pair = [[p, s] for p, s in zip(position, speed)]

    stack = []
    for p, s in sorted(pair)[::-1]: # Reverse Sorted Order
      stack.append((target - p) / s) # (distance - position) / speed
      if len(stack) >= 2 and stack[-1] <= stack[-2]: # car behind is catching up
        stack.pop()
    return len(stack)
```

###***13.Simplify Path (Medium)***
https://leetcode.com/problems/simplify-path/

```python
class Solution:
  def simplifyPath(self, path: str) -> str:
    stack = []
    cur = ""

    for c in path + "/":
      if c == "/": # / is special character
        if cur == "..": # .. go back to previous directory
          if stack: stack.pop()
        elif cur != "" and cur != ".":
          stack.append(cur)
        cur = "" # reset cur
      else:
        cur += c

    return "/" + "/".join(stack)
```

###***14. Decode String (Medium)***
https://leetcode.com/problems/decode-string/

```python
class Solution:
  def decodeString(self, s: str) -> str:
    stack = []

    for i in range(len(s)):
      if s[i] != "]":
        stack.append(s[i])
      else:
        substr = ""
        while stack[-1] != "[":
          substr = stack.pop() + substr # Adding at front to maintain order
        stack.pop() # Popping the [ bracket

        k = ""
        while stack and stack[-1].isdigit():
          k = stack.pop() + k # To maintain order
        stack.append(int(k) * substr)

    return "".join(stack)
```

###***15. Remove K Digits (Medium)***
https://leetcode.com/problems/remove-k-digits/

```python
class Solution:
  def removeKdigits(self, num: str, k: int) -> str:
    stack = []
    for i in num:
      while stack and stack[-1] > i and k > 0: # Pop larget int from stack
        k -= 1
        stack.pop()
      if stack or i != "0": # helps in handling leading zeros
        stack.append(i)
    if k: # 123456 -> 123, if k is still positive such as in increasing order
      stack = stack[:-k]
    return ''.join(stack) or '0'
```

###***16. Remove All Adjacent Duplicates In String II (Medium)***
https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string-ii/

```python
class Solution:
  def removeDuplicates(self, s: str, k: int) -> str:
    stack = []  # [char, count]

    for c in s:
      if stack and stack[-1][0] == c: # If char is equal to c
        stack[-1][1] += 1 # Increase count by one
      else:
        stack.append([c, 1]) # [char, count]

      if stack[-1][1] == k: # Duplicate counts == k
        stack.pop()

    res = ""
    for char, count in stack:
      res += char * count

    return res
```

###***17. 132 Pattern (Medium)***
https://leetcode.com/problems/132-pattern/

```python
from typing import List

class Solution:
  def find132pattern(self, nums: List[int]) -> bool:
    stack = [] # pair [num, minLeft], mono decreasing
    curMin = nums[0]

    for n in nums[1:]:
      while stack and n >= stack[-1][0]:
        stack.pop()
      if stack and n > stack[-1][1]:
        return True

      stack.append([n, curMin])
      curMin = min(curMin, n)
    return False
```


###***18. Maximum Frequency Stack (Hard)***
https://leetcode.com/problems/maximum-frequency-stack/

```python
class FreqStack:

  def __init__(self):
    self.cnt = {} # count of each variable
    self.maxCnt = 0
    self.stacks = {} # stack of stacks

  def push(self, val: int) -> None:
    valCnt = 1 + self.cnt.get(val, 0)
    self.cnt[val] = valCnt
    if valCnt > self.maxCnt:
      self.maxCnt = valCnt
      self.stacks[valCnt] = []
    self.stacks[valCnt].append(val)

  def pop(self) -> int:
    res = self.stacks[self.maxCnt].pop()
    self.cnt[res] -= 1
    if not self.stacks[self.maxCnt]:
      self.maxCnt -= 1
    return res
```

###***19. Largest Rectangle In Histogram (Hard)***
https://leetcode.com/problems/largest-rectangle-in-histogram/

```python
from typing import List

class Solution:
  def largestRectangleArea(self, heights: List[int]) -> int:
    maxArea = 0
    stack = [] # pair: (index, height)

    for i, h in enumerate(heights):
      start = i
      while stack and stack[-1][1] > h:
        index, height = stack.pop()
        maxArea = max(maxArea, height * (i - index))
        start = index
      stack.append((start, h))

    for i, h in stack:
      maxArea = max(maxArea, h * (len(heights) - i))
    return maxArea
```
