###***1. Maximum Subarray (Medium)***
https://leetcode.com/problems/maximum-subarray/

```python
from typing import List

# Sliding window problem, Kadane Algorithm

def max_sub_array(nums: List[int]) -> int:
    max_sub = nums[0]
    current_sum = 0

    for n in nums:
        if current_sum < 0: # current_sum = max(current_sum, 0)
            current_sum = 0
        current_sum += n
        max_sub = max(max_sub, current_sum)
    return max_sub

nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_sub_array(nums))
```

###***2. Maximum Sum Circular Subarray (Medium)***
https://leetcode.com/problems/maximum-sum-circular-subarray/


```python
from typing import List

class Solution:
  def maxSubarraySumCircular(self, nums: List[int]) -> int:
    globMax, globMin = nums[0], nums[0]
    curMax, curMin = 0, 0
    total = 0

    for i, n in enumerate(nums):
      curMax = max(curMax + n, n)
      curMin = min(curMin + n, n)
      total += n
      globMax = max(curMax, globMax)
      globMin = min(curMin, globMin)

    return max(globMax, total - globMin) if globMax > 0 else globMax
```

###***3. Longest Turbulent Array (Medium)***
https://leetcode.com/problems/longest-turbulent-subarray/

```python
from typing import List

class Solution:
  def maxTurbulenceSize(self, arr: List[int]) -> int:
    l, r = 0, 1
    res, prev = 1, ""

    while r < len(arr):
      if arr[r - 1] > arr[r] and prev != ">":
        res = max(res, r - l + 1)
        r += 1
        prev = ">"
      elif arr[r - 1] < arr[r] and prev != "<":
        res = max(res, r - l + 1)
        r += 1
        prev = "<"
      else:
        r = r + 1 if arr[r] == arr[r - 1] else r # [1, 4, 3, 8, 8, 7, 9], can skip only if equal
        l = r - 1
        prev = ""

    return res
```

###***4. Jump Game (Medium)***
https://leetcode.com/problems/jump-game/

```python
from typing import List

class Solution:
  def canJump(self, nums: List[int]) -> bool:
    goal = len(nums) - 1

    for i in range(len(nums) - 1, -1, -1): # check from the end/backward
      if i + nums[i] >= goal: # goal is index of array/not value
        # i + nums[i] -> current index + no jumps available at that index
        goal = i

    return True if goal == 0 else False
```

###***5. Jump Game II (Medium)***
https://leetcode.com/problems/jump-game-ii/

```python
from typing import List

class Solution:
  def jump(self, nums: List[int]) -> int:
    res = 0
    l = r = 0

    while r < len(nums) - 1:
      farthest = 0
      for i in range(l, r + 1): # r + 1 inculsive of r
        farthest = max(farthest, i + nums[i])
        # i + nums[i] -> current index + no jumps available at that index
      l = r + 1
      r = farthest
      res += 1
    return res
```

###***6. Jump Game VII (Medium)***
https://leetcode.com/problems/jump-game-vii/

```python
class Solution:
  def canReach(self, s: str, minJump: int, maxJump: int) -> bool:
    q, farthest = deque([0]), 0

    while q:
      i = q.popleft()
      start = max(i + minJump, farthest + 1) # farthest is the point we already reached
      for j in range(start, min(i + maxJump + 1, len(s))):
        if s[j] == "0":
          q.append(j)
          if j == len(s) - 1:
            return True
      farthest = i + maxJump
    return False
```

###***7. Gas Station (Medium)***
https://leetcode.com/problems/gas-station/

```python
from typing import List

class Solution:
  def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
    # total gas must be equal or more than cost to finish looping
    if sum(gas) < sum(cost):
      return -1

    total = 0
    res = 0 # starting position
    for i in range(len(gas)):
      total += (gas[i] - cost[i])
      if total < 0:
        total = 0
        res = i + 1 # next starting position

    return res
```

###***8. Hand of Straights (Medium)***
https://leetcode.com/problems/hand-of-straights/

```python
from typing import List

class Solution:
  def isNStraightHand(self, hand: List[int], groupSize: int) -> bool:
    if len(hand) % groupSize:
      return False

    count = {}
    for n in hand:
      count[n] = 1 + count.get(n, 0) # Counter can also be used

    minH = list(count.keys()) # min heap works with list
    heapq.heapify(minH)

    while minH:
      first = minH[0] # min heap index 0 gives as minimum value

      for i in range(first, first + groupSize):
        if i not in count:
          return False
        count[i] -= 1
        if count[i] == 0:
          if i != minH[0]: # key trick
            return False
          heapq.heappop(minH)

    return True
```

###***9. Minimize Maximum of Array (Medium)***
https://leetcode.com/problems/minimize-maximum-of-array/

```python
from typing import List

class Solution:
  def minimizeArrayValue(self, nums: List[int]) -> int:
    res = nums[0]
    total = nums[0]

    for i in range(1, len(nums)):
      total += nums[i]
      res = max(res, math.ceil(total / (i + 1)))
    return res
```

###***10. Dota2 Senate (Medium)***
https://leetcode.com/problems/dota2-senate/

```python
class Solution:
  def predictPartyVictory(self, senate: str) -> str:
    senate = list(senate)
    D, R = deque(), deque()

    for i, c in enumerate(senate):
      if c == "R":
        R.append(i)
      else:
        D.append(i)

    while D and R:
      dTurn = D.popleft()
      rTurn = R.popleft()

      if rTurn < dTurn: # Smaller index val, the one that comes first
        R.append(dTurn + len(senate)) # Adding offest for looping
      else:
        D.append(rTurn + len(senate))

    return "Radiant" if R else "Dire"
```

###***11. Maximum Points You Can Obtain From Cards (Medium)***
https://leetcode.com/problems/maximum-points-you-can-obtain-from-cards/

```python
from typing import List

class Solution:
  def maxScore(self, cardPoints: List[int], k: int) -> int:
    l, r = 0, len(cardPoints) - k
    total = sum(cardPoints[r:])
    res = total

    while r < len(cardPoints): # Sliding window
      total += (cardPoints[l] - cardPoints[r]) # Add left val and remove rightmost val
      res = max(res, total)
      l += 1
      r += 1

    return res
```

###***12. Merge Triplets to Form Target Triplet (Medium)***
https://leetcode.com/problems/merge-triplets-to-form-target-triplet/

```python
from typing import List

class Solution:
  def mergeTriplets(self, triplets: List[List[int]], target: List[int]) -> bool:
    good = set()

    for t in triplets:
      if t[0] > target[0] or t[1] > target[1] or t[2] > target[2]:
        continue

      for i, v in enumerate(t):
        if v == target[i]:
          good.add(i)

    return len(good) == 3
```

###***13. Partition Labels (Medium)***
https://leetcode.com/problems/partition-labels/

```python
from typing import List

class Solution:
  def partitionLabels(self, s: str) -> List[int]:
    lastIndex = {} # char -> last index in s

    for i, c in enumerate(s):
      lastIndex[c] = i # last index will be updated with loop

    res = []
    size, end = 0, 0
    for i, c in enumerate(s):
      size += 1
      end = max(end, lastIndex[c])

      if i == end:
        res.append(size)
        size = 0 # new partition will start
      # end doesn't need to be updated here since max is taken every iteration

    return res
```

###***14. Valid Parenthesis String (Medium)***
https://leetcode.com/problems/valid-parenthesis-string/

```python
from typing import List

class Solution:
  def checkValidString(self, s: str) -> bool:
    leftMin, leftMax = 0, 0

    for c in s:
      if c == "(":
        leftMin, leftMax = leftMin + 1, leftMax + 1
      elif c == ")":
        leftMin, leftMax = leftMin - 1, leftMax - 1
      else:
        leftMin, leftMax = leftMin - 1, leftMax + 1

      if leftMax < 0:
        return False
      if leftMin < 0: # for this case -> s = ( * ) (
        leftMin = 0

    return leftMin == 0
```

###***15. Eliminate Maximum Number of Monsters (Medium)***
https://leetcode.com/problems/eliminate-maximum-number-of-monsters/

```python
from typing import List

class Solution:
  def eliminateMaximum(self, dist: List[int], speed: List[int]) -> int:
    minReach = [] # minute required for monster to reach city
    for d, s in zip(dist, speed):
      minute = math.ceil(d / s) # time = distance/speed
      minReach.append(minute)

    minReach.sort() # fastest montster to reach the city will be eliminated first
    res = 0
    for minute in range(len(minReach)): # minute is index here
      if minute >= minReach[minute]: # current monster can't be eliminated
        return res
      res += 1
    return res
```

###***16. Two City Scheduling (Medium)***
https://leetcode.com/problems/two-city-scheduling/

```python
from typing import List

class Solution:
  def twoCitySchedCost(self, costs: List[List[int]]) -> int:
    diffs = []
    for c1, c2 in costs:
      diffs.append([c2 - c1, c1, c2]) # diff, city1, city2
    diffs.sort()
    res = 0
    for i in range(len(diffs)):
      if i < len(diffs) / 2: # city2 will be on left side
        res += diffs[i][2]
      else:
        res += diffs[i][1] # city1 will be on right side
    return res
```

###***17. Maximum Length of Pair Chain (Medium)***
https://leetcode.com/problems/maximum-length-of-pair-chain/

```python
from typing import List

class Solution:
  def findLongestChain(self, pairs: List[List[int]]) -> int:
    pairs.sort(key=lambda p: p[1]) # Sort by second item/ end time
    length = 1
    end = pairs[0][1] # First pair, ending val
    for i in range(1, len(pairs)): # Skip first pair
      if end < pairs[i][0]: # Start of i comes after end
        length += 1
        end = pairs[i][1] # Update the ending

    return length
```

###***18. Minimum Deletions to Make Character Frequencies Unique (Medium)***
https://leetcode.com/problems/minimum-deletions-to-make-character-frequencies-unique/

```python
class Solution:
  def minDeletions(self, s: str) -> int:
    count = collections.Counter(s)
    used_freq = set()
    res = 0 # total no of deletions

    for c, freq in count.items():
      while freq > 0 and freq in used_freq: # freq > 0, edge case
        freq -= 1 # frequency is reduced
        res += 1 # no deletion is increased
      used_freq.add(freq) # freq = either unique or 0
    return res
```

###***19. Candy (Hard)***
https://leetcode.com/problems/candy/

```python
class Solution:
  def candy(self, ratings: List[int]) -> int:
    candies = [1] * len(ratings) # Initialize with 1

    # Traverse from left to right
    for i in range(1, n): # Skipping 1st index, no left neighbor
      if ratings[i] > ratings[i - 1]: # only if i is greater than left neighbor
        candies[i] = candies[i - 1] + 1

    # Traverse from right to left
    for i in range(n - 2, -1, -1): # Skipping last index, no right neighbor
      if ratings[i] > ratings[i + 1]: # i is greater than right neighbor
        candies[i] = max(candies[i], candies[i + 1] + 1) # take max val

    return sum(candies)
```
